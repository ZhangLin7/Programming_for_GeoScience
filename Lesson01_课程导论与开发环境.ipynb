## 一、Jupyter Notebook 极简使用指南

在动手写代码之前，先花2分钟认识一下你正在使用的工具。

### Jupyter 是什么？

Jupyter Notebook 是一个**交互式编程环境**——

| **Markdown Cell** | 写文字、说明、标题 | 普通文本格式（你现在读的这段就是） |

### 最最最常用的快捷键（请记住）

- **`A`** —— 在当前 cell 上方插入新 cell（需先按 Esc）
- **`B`** —— 在当前 cell 下方插入新 cell（需先按 Esc）
- **`DD`**（连按两次D）—— 删除当前 cell

### 3.2 试一试：算一道地学题

**情境**：地球的平均半径约 6371 公里，地球周长是多少公里？

**公式**：周长 = 2 × π × 半径

我们用 Python 来算：

````python
# 地球周长 = 2 × π × 半径
# 这里我们用近似的 π = 3.14159

print(2 * 3.14159 * 6371)
````

## 五、看看真正的地学Python能做什么

下面这段代码现在你**完全看不懂**没关系——这是我们40学时之后大家就能写出来的东西。

现在只是让你感受一下：**真正的地学Python代码长什么样**。

（这段代码**不需要你运行**，只是给你看一眼。）

```python
# ===== 这是 40 学时后你能写出来的代码（仅展示，无需运行）=====

import rasterio
import numpy as np
import matplotlib.pyplot as plt

# 读取 Landsat 卫星影像
with rasterio.open('landsat.tif') as src:
    red = src.read(4).astype('float32')   # 红波段
    nir = src.read(5).astype('float32')   # 近红外波段

# 计算 NDVI（归一化植被指数）—— 一行公式就能识别植被
ndvi = (nir - red) / (nir + red)

# 显示结果
plt.imshow(ndvi, cmap='RdYlGn')
plt.colorbar()
plt.title('NDVI 植被指数图')
plt.show()

print(f'植被覆盖率：{(ndvi > 0.4).sum() / ndvi.size * 100:.1f}%')
```

**这段代码做了什么？**

- 读取了一景卫星影像
- 计算了归一化植被指数
- 画出了彩色专题图
- 统计了植被覆盖率

**总共：不到20行代码。**
