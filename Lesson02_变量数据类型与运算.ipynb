## 一、什么是变量？

### 1.1 一个让你头疼的场景

假设我们要算地球周长，公式是 `2 × π × 半径`。

上次课我们这样写：

````python
# 计算地球周长
print(2 * 3.14159 * 6371)
````

现在问题来了——我还想算地球的**表面积**（公式 `4 × π × 半径²`）和**体积**（公式 `(4/3) × π × 半径³`）：

````python
# 计算地球周长
print(2 * 3.14159 * 6371)

# 计算地球表面积
print(4 * 3.14159 * 6371 * 6371)

# 计算地球体积
print((4/3) * 3.14159 * 6371 * 6371 * 6371)
````

看出问题了吗？

- 半径 `6371` 写了 **6 次**
- π `3.14159` 写了 **3 次**
- 如果将来要把半径改成更精确的 `6371.393`，要改 6 个地方，**漏改一个就出bug**

**有没有办法让数据'有名字'，用一次就能复用？** 这就是变量。

### 1.2 变量：给数据起一个名字

**变量（Variable）** 就是给一个数据起的名字。

用 `=` 把一个值"赋"给变量名：

````python
# 定义两个变量
radius = 6371      # 把数字 6371 赋给变量 radius
pi = 3.14159       # 把数字 3.14159 赋给变量 pi

# 现在可以反复使用它们了
print(2 * pi * radius)            # 周长
print(4 * pi * radius * radius)   # 表面积
````

**这就是变量的威力**：

- 数据有了"名字"，代码读起来更直观（`pi * radius` 比 `3.14159 * 6371` 清楚得多）
- 想改数值只需要改一处（把半径改成更精确的值，只改第一行）
- 同一个数据可以反复使用

### 1.3 变量可以重新赋值

变量名是一个'盒子'，里面装的东西可以换：

````python
city = "杭州"
print(city)

city = "北京"   # 重新赋值，原来的'杭州'被'北京'替换了
print(city)

city = "塞罕坝"  # 再来一次
print(city)
````

### 1.4 变量命名规则（必须遵守）

**强制规则**（违反会报错）：

1. ✅ 只能由**字母、数字、下划线**组成；
2. ✅ **不能以数字开头**；
3. ✅ **区分大小写**（`Radius` 和 `radius` 是两个不同的变量）；
4. ❌ 不能使用 Python 的**关键字**（如 `print`、`if`、`for`、`in` 等）。

**良好习惯**（不强制，但强烈建议）：

5. 变量名要**有意义**——`radius` 比 `r` 好，`r` 比 `x` 好；
6. 多个单词用**下划线连接**——`elevation_max`、`city_name`；
7. 不要用**中文**作变量名——虽然技术上可以，但不专业。

````python
# ✅ 好的变量名
city_name = "杭州"
elevation_max = 8848.86
is_china = True

print(city_name, elevation_max, is_china)
````

````python
# ❌ 不好的变量名（虽然能跑，但不推荐）
x = "杭州"          # 看不出x是什么
a = 8848.86         # 不知道这是什么数
城市 = "杭州"        # 中文，不专业

# ❌ 这些会直接报错（请勿运行）
# 1city = "杭州"     # 不能以数字开头
# city-name = "杭州"  # 不能用减号
# print = 100        # 覆盖了内置函数 print，灾难
````

## 二、Python 的基本数据类型

上面的练习中，你可能注意到：海拔是数字、站名是文字、是否境内是 True。这些都是**数据类型不同**。

Python 有四种最基本的数据类型，我们今天都要认识：

| 类型名 | 全称 | 含义 | 例子 |
|---|---|---|---|
| `int` | integer | 整数 | `100`、`-5`、`0`、`6371` |
| `float` | floating point | 浮点数（小数） | `3.14`、`8848.86`、`-0.5` |
| `str` | string | 字符串（文字） | `"杭州"`、`'GIS'`、`"123"` |
| `bool` | boolean | 布尔值（真/假） | `True`、`False` |

### 2.1 用 type() 函数查看类型

Python 提供了一个内置函数 `type()`，可以查看任何变量的类型：

````python
# 看看四种基本类型
elevation = 6371          # 整数
pi = 3.14159              # 浮点数
city = "杭州"             # 字符串
is_china = True           # 布尔值

print(type(elevation))
print(type(pi))
print(type(city))
print(type(is_china))
````

### 2.2 整型 int 与浮点型 float

整数和浮点数的**区别只在于有没有小数点**：

- `100` 是 int
- `100.0` 是 float（哪怕小数部分是0）

````python
a = 100
b = 100.0

print(type(a))   # int
print(type(b))   # float

# 但是它们的值是相等的
print(a == b)    # True
````

````python
# 数学运算
print(10 + 3)      # 13       int + int = int
print(10 - 3)      # 7
print(10 * 3)      # 30
print(10 / 3)      # 3.3333... 注意！除法结果总是 float
print(10 // 3)     # 3        整除（向下取整）
print(10 % 3)      # 1        取余数
print(10 ** 3)     # 1000     幂运算（10的3次方）
````

**重点提醒**（很多同学第18次课会踩坑）：

- `/`（除法）：结果总是 **float**，例如 `10/3 = 3.333...`，`10/2 = 5.0`
- `//`（整除）：结果舍弃小数，例如 `10//3 = 3`
- 这两个符号一定要区分清楚

### 2.3 布尔型 bool —— 只有 True 和 False

布尔值用来表示**真/假**两种状态。注意 **首字母必须大写**！

````python
is_in_china = True
is_water = False

print(is_in_china)
print(is_water)
print(type(is_in_china))
````

````python
# 比较运算的结果就是布尔值
elevation = 8848
print(elevation > 8000)       # True
print(elevation == 8848)      # True
print(elevation < 5000)       # False

# 注意：判断相等用 == （两个等号），不是 = （一个等号）
# 一个等号是赋值，两个等号是比较
````

**踩坑预警**：

❌ `True` 写成 `true` —— Python 区分大小写，`true` 会被当成变量名而报错

❌ 比较相等用 `=` —— `=` 是赋值，`==` 才是比较

## 三、字符串 str 详解

字符串就是"文字"——用引号包起来的内容。在地学领域，地名、坐标字符串、文件路径、属性值等都是字符串。

### 3.1 字符串的三种写法

````python
# 写法1：单引号
s1 = '杭州'

# 写法2：双引号
s2 = "杭州"

# 写法3：三引号（用于多行字符串）
s3 = """杭州市，简称"杭"，
古称临安、钱塘，
浙江省辖地级市。"""

print(s1)
print(s2)
print(s3)
````

**单引号和双引号是等价的**——选哪个看自己习惯。但有一个小技巧：

- 如果文字里有双引号，外面就用单引号：`'他说"你好"'`
- 如果文字里有单引号，外面就用双引号：`"It's GIS"`

### 3.2 字符串的常用操作

````python
city = "杭州"
province = "浙江"

# 1. 拼接：用 +
full_name = province + city
print(full_name)        # 浙江杭州

# 2. 复制：用 *
line = "=" * 30
print(line)             # ==============================

# 3. 长度：用 len() 函数
print(len(city))        # 2
print(len("GIS"))       # 3

# 4. 字符串相加（注意，数字也可以）
code = "330100" + ""
print(code, type(code)) # 330100 str
````

````python
# 一个有用的小例子：用字符串运算画一个标题框
title = "杭州市基本信息"
border = "=" * 30

print(border)
print(title)
print(border)
````

### 3.3 字符串的索引（先了解，下节课详细学）

字符串里的每个字符都有一个**位置编号（索引）**，从 0 开始：

````python
s = "Hello GIS"
#    01234567 8
#    第0个是H，第1个是e，第8个是S

print(s[0])     # H
print(s[6])     # G
print(s[-1])    # 最后一个：S

# 详细的字符串处理我们下次课再讲
````

## 四、类型转换

数据类型之间可以相互转换——这是地学编程中**非常常用**的操作。

### 4.1 三个核心转换函数

| 函数 | 作用 | 例子 |
|---|---|---|
| `int(x)` | 转成整数 | `int("100")` → `100` |
| `float(x)` | 转成浮点数 | `float("3.14")` → `3.14` |
| `str(x)` | 转成字符串 | `str(100)` → `"100"` |

````python
# 字符串 → 数字
elevation_str = "8848"           # 这是一个字符串
elevation_num = int(elevation_str) # 转换成整数

print(elevation_str, type(elevation_str))   # 8848 str
print(elevation_num, type(elevation_num))   # 8848 int

# 现在可以做数学运算了
print(elevation_num + 1)        # 8849
````

````python
# 数字 → 字符串（拼接的时候必须做这一步！）
elevation = 8848

# ❌ 这样会报错——字符串和数字不能直接相加
# print("珠峰高度：" + elevation + "米")

# ✅ 必须先转成字符串
print("珠峰高度：" + str(elevation) + "米")
````

### 4.2 ⚠️ 重要：input() 永远返回字符串

Python 有一个 `input()` 函数可以让用户输入数据。**注意**：input() 不管你输入什么数字，**返回的永远是字符串**！

````python
# 这段代码运行后，请输入一个数字（比如100）
user_input = input("请输入一个海拔值（米）：")

print(user_input, type(user_input))    # 即使输入的是数字，类型也是 str！
````

````python
# 想做数学运算，必须先转换
user_input = input("请输入一个海拔值（米）：")
elevation = float(user_input)   # 把字符串转成浮点数

print("输入的海拔：", elevation, type(elevation))
print("加100后：", elevation + 100)
````

**记住这个铁律**：用 input() 接收的数据，几乎一定要做类型转换！

### 4.3 常见的类型转换坑

**坑1**：字符串里如果有非数字字符，会报错

````python
# 这一行会报错（ValueError）
# int("8848米")     # ❌ 因为 "米" 不是数字

# 正确做法：先去掉非数字部分（下次课学字符串处理时讲）
# 现在你只需要知道：要转换的字符串必须是"纯数字"

print(int("8848"))      # ✅ 8848
````

**坑2**：浮点数转整数会**截断**小数（不是四舍五入）

````python
print(int(8848.86))     # 8848（不是8849！）
print(int(8848.99))     # 8848（截断，不是四舍五入）
print(int(-3.7))        # -3
````

如果想四舍五入，用 `round()` 函数：

````python
print(round(8848.86))      # 8849
print(round(8848.49))      # 8848
print(round(8848.86, 1))   # 8848.9（保留1位小数）
````

## 五、f-string：最优雅的格式化输出（重点！）

前面我们用 `+` 拼接字符串，但是当变量多了，代码会很难看：

```python
print("城市：" + city + "，海拔：" + str(elevation) + "米，是否在中国：" + str(is_china))
```

Python 提供了一种更优雅的方式 —— **f-string**。

### 5.1 f-string 的基本写法

**写法**：在字符串前面加一个字母 `f`，然后在 `{}` 里写变量名：

````python
city = "杭州"
elevation = 41.7

# 老方法：拼接
print("城市：" + city + "，平均海拔：" + str(elevation) + "米")

# 新方法：f-string
print(f"城市：{city}，平均海拔：{elevation}米")

# 是不是优雅多了？
````

**对比一下**：

- 老方法：`"..." + str(变量) + "..."` —— 麻烦、容易忘 str()
- f-string：`f"...{变量}..."` —— 简洁，自动处理类型转换

**强烈推荐**：从今天起，所有需要把变量嵌入字符串的场景，**都用 f-string**。

### 5.2 f-string 里可以做计算

````python
radius = 6371
pi = 3.14159

# {} 里不仅能放变量，还能直接做计算
print(f"地球周长：{2 * pi * radius} 公里")
print(f"地球半径的两倍：{radius * 2} 公里")
````

### 5.3 ✨ 控制小数位数（极其常用）

地学计算结果常常有很多位小数，看起来很乱。f-string 可以一行控制小数位数：

**语法**：`{变量:.N f}` 表示保留 N 位小数

````python
radius = 6371
pi = 3.14159
circumference = 2 * pi * radius

print(f"原始结果：{circumference}")
print(f"保留 2 位小数：{circumference:.2f}")
print(f"保留 0 位小数：{circumference:.0f}")
print(f"保留 4 位小数：{circumference:.4f}")
````

````python
# 一个完整的地学应用例子
city = "珠穆朗玛峰大本营"
longitude = 86.92345
latitude = 27.98765
elevation = 5364.789

# 用 f-string 做一份漂亮的'信息卡'
print("=" * 40)
print(f"地点：{city}")
print(f"经度：{longitude:.2f}°E")
print(f"纬度：{latitude:.2f}°N")
print(f"海拔：{elevation:.1f} 米")
print("=" * 40)
````

### 5.4 f-string 的更多技巧（了解即可）

下面这些先了解，将来用到再回看：

````python
value = 0.456789

# 转百分比（常用！）
print(f"植被覆盖率：{value:.2%}")    # 45.68%

# 千位分隔符（金额、人口等大数字常用）
population = 1413426853
print(f"中国人口：{population:,} 人")    # 1,413,426,853 人

# 字符串对齐（做表格输出常用）
name = "杭州"
print(f"|{name:<10}|")    # 左对齐10字符
print(f"|{name:>10}|")    # 右对齐10字符
print(f"|{name:^10}|")    # 居中10字符
````

### 课后作业 3：思考题参考答案

**问题**：如果有 `s = "3.14abc"`，怎么提取出 3.14？

**思路**（多种）：

1. **字符串方法**：用 `s.replace("abc", "")` 把 "abc" 替换为空字符串，得到 "3.14"，再 float() 转换；

2. **字符串切片**：用索引取前4个字符 `s[:4]`，得到 "3.14"，再 float() 转换（但这要求你**预先知道**数字部分长度）；

3. **正则表达式**（更通用，将来高级方法）：用正则匹配数字部分；

4. **逐字符遍历**（基础方法）：循环每个字符，判断是否是数字或小数点。

**这些方法的共同点**：都需要**字符串处理技术**——这正是下节课的内容。这个思考题的目的就是让你意识到"现有工具不够用"，从而对下节课产生期待。
