## 一、比较运算符 —— 让程序'问问题'

在条件判断之前，我们先回顾一下'比较'。

### 1.1 六个比较运算符

| 运算符 | 含义 | 例子 | 结果 |
|---|---|---|---|
| `==` | 等于 | `8848 == 8848` | `True` |
| `!=` | 不等于 | `8848 != 5000` | `True` |
| `>` | 大于 | `8848 > 5000` | `True` |
| `<` | 小于 | `8848 < 5000` | `False` |
| `>=` | 大于等于 | `8848 >= 8848` | `True` |
| `<=` | 小于等于 | `5000 <= 8848` | `True` |

**比较的结果永远是布尔值（`True` 或 `False`）。**

````python
# 跟着敲一下
elevation = 8848

print(elevation == 8848)    # True
print(elevation != 5000)    # True
print(elevation > 5000)     # True
print(elevation < 5000)     # False
print(elevation >= 8848)    # True
print(elevation <= 5000)    # False
````

### 1.2 ⚠️ 最最重要的一对：= 和 ==

**这是零基础学生最高频混淆的两个符号**：

| 符号 | 名字 | 用途 | 例子 |
|---|---|---|---|
| `=` | 一个等号 | **赋值**（把右边的值放进左边的盒子） | `x = 100` |
| `==` | 两个等号 | **比较**（判断左右是否相等） | `x == 100` |

**记忆口诀**：一个等号是"装入"，两个等号是"比较"。

````python
# 跟着敲，体会差异
x = 100              # = 是赋值：把100放进x
print(x)             # 100

result = x == 100    # == 是比较：判断x是否等于100
print(result)        # True
````

### 1.3 字符串也能比较

比较运算符不只能比较数字，**字符串也能比较**：

````python
direction = "N"

print(direction == "N")    # True
print(direction == "S")    # False
print(direction != "N")    # False

# ⚠️ 字符串比较是严格的，大小写不一样就不相等
print("GIS" == "gis")      # False
````

## 二、if 单分支 —— 让程序'有条件地做事'

### 2.1 最简单的 if 语法

**🔑 语法**：

```python
if 条件:
    要执行的代码
```

**意思**：如果条件为 `True`，就执行下面缩进的代码；如果为 `False`，就跳过。

**三个细节请特别注意**：

1. `if` 后面跟一个**条件表达式**（结果是 True 或 False）
2. 条件后面有一个**冒号 `:`**（不能漏！）
3. 下一行的代码必须**缩进**（通常缩进4个空格，**Jupyter 会自动帮你缩进**）

````python
# 第一个if例子：判断珠峰是否够高
elevation = 8848

if elevation > 8000:
    print("这是一座超高峰！")

print("程序结束")
````

````python
# 改一下数字试试：5000 米呢？
elevation = 5000

if elevation > 8000:
    print("这是一座超高峰！")    # 这一行不会执行

print("程序结束")
````

**观察**：当条件为 `False` 时，缩进的那一行被**跳过了**，但 `print("程序结束")` 因为**没有缩进**，所以照常执行。

👉 **这就是缩进的意义——它告诉 Python：哪些代码属于 if、哪些不属于。**

### 2.2 ⭐ 缩进规则 —— 本节课最大的难点

**Python 用'缩进'表示代码的层次关系**，这跟很多其他语言（比如C、Java用花括号）完全不同。

**三条铁律**：

1. **同一层的代码必须缩进相同的空格数**（推荐4个空格）
2. **缩进的代码属于上一行的'块'**（比如 if 块）
3. **缩进不一致 = 报错（IndentationError）**

````python
# ✅ 正确：缩进一致
elevation = 8848

if elevation > 8000:
    print("超高峰")           # 4个空格缩进
    print("应进行特殊保护")    # 也是4个空格，与上面对齐
````

````python
# ❌ 错误：缩进不一致 —— 这段代码会报错（请勿运行，了解即可）
# if elevation > 8000:
#     print("超高峰")           # 4个空格
#       print("应进行特殊保护")   # 6个空格 → IndentationError
````

**📌 缩进的小贴士**：

- **Jupyter Notebook 会自动缩进**——你在 `if 条件:` 后回车，下一行自动缩进 4 个空格
- **不要混用 Tab 和空格**——这是初学者最容易的隐形错误。统一用 4 个空格最安全
- **如果出现 IndentationError**——99% 是因为缩进数量不一致，仔细看每行前面的空格数

### 2.3 if 块里可以有多行代码

````python
# 一个 if 块可以包含多行（每行都要缩进）
ndvi = 0.65

if ndvi > 0.4:
    print("该区域有植被")
    print(f"NDVI 值：{ndvi}")
    print("建议保护")

print("分析完成")    # 这行没缩进，无论 if 是否执行，它都会执行
````

## 三、if-else 二分支 —— 二选一

如果条件为 True 做一件事，**False 时做另一件事**——这就是 if-else 结构。

### 3.1 if-else 语法

```python
if 条件:
    条件成立时执行
else:
    条件不成立时执行
```

**注意**：`else` 后面**也有冒号**，下面的代码也要缩进。

````python
# 经典例子：判断点是否在研究区内
longitude = 116.40

# 假设研究区是经度 [73, 135] 之间（中国大陆经度范围简化）
if longitude >= 73 and longitude <= 135:
    print("该点在研究区内")
else:
    print("该点在研究区外")
````

````python
# 把 longitude 改成 -120 试试
longitude = -120

if longitude >= 73 and longitude <= 135:
    print("该点在研究区内")
else:
    print("该点在研究区外")
````

### 3.2 ✨ 结构图：if-else 的执行流程

```
        ┌──────────┐
        │  条件?   │
        └────┬─────┘
    True ←──┴──→ False
      ↓           ↓
  if 块代码    else 块代码
      ↓           ↓
      └──→ 继续执行 ←──┘
```

**关键观念**：if 和 else **永远只有一个会执行**，绝对不会两个都执行。

## 四、if-elif-else 多分支 —— 多选一

现实中，很多情况是**两个以上的可能性**——这就需要多分支。

### 4.1 if-elif-else 语法

```python
if 条件1:
    条件1成立时执行
elif 条件2:
    条件1不成立但条件2成立时执行
elif 条件3:
    前面都不成立，条件3成立时执行
else:
    前面所有条件都不成立时执行
```

- `elif` 是 "else if" 的缩写——"否则如果"
- `elif` 可以有任意多个
- `else` 最多一个，且必须在最后（也可以省略）
- **从上到下依次判断，遇到第一个 True 就执行对应代码块，然后跳出整个结构**

````python
# 经典例子：根据温度判断气候等级
temperature = 25

if temperature < 0:
    print("严寒")
elif temperature < 10:
    print("寒冷")
elif temperature < 20:
    print("凉爽")
elif temperature < 30:
    print("温暖")
else:
    print("炎热")
````

**重要思考**：上面 `temperature = 25` 时，输出的是 `温暖`。

为什么不是 `寒冷`（25 < 10 是 False，但 25 也 ">= 0"，是不是也算 "不严寒"？）？

**答**：因为 elif 是**互斥的**——前面的条件没满足才会判断后面。也就是说：
- 第一个 elif 的真实含义是："`temperature >= 0` 且 `temperature < 10`" 
- 第二个 elif 的真实含义是："`temperature >= 10` 且 `temperature < 20`"
- ……

**这就是为什么我们写 elif 时只需要写右边界（< 30）即可，左边界自动由前面的条件决定。**

### 4.2 🌟 经典地学应用：NDVI 五级分类

还记得第18次课要做的 NDVI 计算吗？分类规则是这样的：

| NDVI 范围 | 地物类型 |
|---|---|
| < 0 | 水体 |
| 0 ~ 0.2 | 裸土/建成区 |
| 0.2 ~ 0.4 | 稀疏植被 |
| 0.4 ~ 0.7 | 中等植被 |
| ≥ 0.7 | 茂密森林 |

**用 if-elif-else 完美实现**：

````python
# NDVI 五级分类
ndvi = 0.55

if ndvi < 0:
    land_type = "水体"
elif ndvi < 0.2:
    land_type = "裸土/建成区"
elif ndvi < 0.4:
    land_type = "稀疏植被"
elif ndvi < 0.7:
    land_type = "中等植被"
else:
    land_type = "茂密森林"

print(f"NDVI = {ndvi} → 地物类型：{land_type}")
````

**💡 这段代码的价值**：第18次课我们要对一整景遥感影像做这种分类——但所有像元都是**同一套判断逻辑**。今天学的 if-elif-else 就是那个分类的核心。

## 五、逻辑运算符 —— 多个条件组合

现实中我们常常需要判断**多个条件同时成立**——比如"经度在中国范围**且**纬度也在中国范围"。这就需要逻辑运算符。

### 5.1 三个逻辑运算符

| 运算符 | 含义 | 真假规则 |
|---|---|---|
| `and` | 与 | 两边都为 True 才是 True |
| `or` | 或 | 两边任一为 True 就是 True |
| `not` | 非 | 取反：True → False、False → True |

````python
# and：两边都要 True
print(True and True)     # True
print(True and False)    # False
print(False and False)   # False
````

````python
# or：任一为 True 就行
print(True or True)      # True
print(True or False)     # True
print(False or False)    # False
````

````python
# not：取反
print(not True)          # False
print(not False)         # True
````

### 5.2 🎯 经典地学应用：判断点是否在矩形研究区内

**问题**：给定一个点的经纬度，判断它是否落在某个矩形研究区内。

假设研究区范围：经度 `[73, 135]`、纬度 `[18, 54]`（中国大陆简化范围）。

**判定条件**：
- 经度 ≥ 73 **且** 经度 ≤ 135
- 纬度 ≥ 18 **且** 纬度 ≤ 54
- 上面**两个条件都要满足**——用 `and` 连接

````python
# 北京：经度 116.40、纬度 39.90
longitude = 116.40
latitude = 39.90

if longitude >= 73 and longitude <= 135 and latitude >= 18 and latitude <= 54:
    print("该点在研究区内")
else:
    print("该点在研究区外")
````

### 5.3 ⚠️ 常见陷阱：`73 <= x <= 135` 这种链式比较

Python 还允许这样写**链式比较**：

````python
# 链式比较（Python 特有的写法）
longitude = 116.40

# 这两种写法效果一样
result1 = 73 <= longitude <= 135
result2 = longitude >= 73 and longitude <= 135

print(result1)    # True
print(result2)    # True
````

**链式比较看起来更优雅**，但**很多其他语言（比如 C、Java）不支持**。Python 是少数支持这种写法的语言。**今天学一个写法就行，看到都要能认出来。**

### 5.4 ⚠️ 高频踩坑：`x == 1 or 2`

**这是一个让所有零基础学生都踩过的坑**：

````python
direction = "N"

# ❌ 错误写法（语法不会报错，但逻辑错了）
if direction == "S" or "W":
    print("南半球或西半球")
# 这个 if 永远是 True！为什么？
# 因为 "or 'W'" 的 'W' 是非空字符串，恒为 True
````

## 六、嵌套条件 —— if 里面套 if

if 块里也可以再放 if，叫**嵌套条件**。

````python
# 例：先判断海拔区间，再判断植被状况
elevation = 2500
ndvi = 0.6

if elevation < 3000:
    print("低海拔区域")
    if ndvi > 0.4:                 # 第二层 if，需要再缩进4个空格
        print("  且植被茂密")        # 第三层缩进
    else:
        print("  且植被稀疏")
else:
    print("高海拔区域")
````

**📌 嵌套要点**：

1. 每多一层嵌套，**缩进多一级**（再加 4 个空格）
2. 嵌套不要太深——超过 3 层基本就难读了，应该改用 `and` 合并条件
3. **绝大多数情况，用 `and` 比嵌套更优雅**：

```python
# 嵌套写法（多层缩进）
if elevation < 3000:
    if ndvi > 0.4:
        print("低海拔且植被茂密")

# 等价的 and 写法（更优雅）
if elevation < 3000 and ndvi > 0.4:
    print("低海拔且植被茂密")
```

````python
# ===== 即时练习③ 综合实战 参考答案 =====
coord_str = "39°54'27\"S"

# Step 1: 取方向字符
direction = coord_str[-1]

# Step 2: 去掉方向字符，只保留数字部分
main_part = coord_str[:-1]

# Step 3: 解析度分秒（沿用上次课的方法）
parts = main_part.replace("°", " ").replace("'", " ").replace("\"", " ").split()
degrees = int(parts[0])
minutes = int(parts[1])
seconds = int(parts[2])

# Step 4: 计算十进制度
decimal = degrees + minutes / 60 + seconds / 3600

# Step 5: 根据方向决定正负号
if direction == "S" or direction == "W":
    decimal = -decimal

# Step 6: 输出
print(f"{coord_str} → {decimal:.4f}°")
````

````python
# ===== 课后作业 3：NDVI 五级分类完整实现 =====

ndvi = 0.55

# 1. 地物类型
if ndvi < 0:
    land_type = "水体"
elif ndvi < 0.2:
    land_type = "裸土/建成区"
elif ndvi < 0.4:
    land_type = "稀疏植被"
elif ndvi < 0.7:
    land_type = "中等植被"
else:
    land_type = "茂密森林"

# 2. 生态价值
if land_type == "水体" or land_type == "茂密森林":
    ecological_value = "高"
elif land_type == "中等植被":
    ecological_value = "中"
else:
    ecological_value = "低"

# 3. 建议措施
if land_type == "水体":
    measure = "水文监测"
elif land_type == "茂密森林":
    measure = "重点保护"
elif land_type == "中等植被":
    measure = "定期监测"
elif land_type == "稀疏植被":
    measure = "植被恢复"
else:
    measure = "土壤改良"

# 4. 输出报告
border = "=" * 30
print(border)
print("NDVI 分析报告")
print(border)
print(f"NDVI 值：{ndvi}")
print(f"地物类型：{land_type}")
print(f"生态价值：{ecological_value}")
print(f"建议措施：{measure}")
print(border)
````
