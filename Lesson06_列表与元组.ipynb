## 一、列表的创建与访问

### 1.1 什么是列表？

**列表（list）**：一个有序的、可改变的元素集合。

**特点**：
- 用方括号 `[]` 创建
- 元素之间用逗号 `,` 分隔
- 元素可以是任何类型（数字、字符串、布尔、甚至列表）
- 元素可以**重复**
- 元素**有顺序**（每个元素都有一个位置编号，从0开始）

````python
# 1. 不同类型的列表
elevations = [2530, 1820, 4200, 3650]              # 数字列表
cities = ["杭州", "北京", "上海"]                    # 字符串列表
is_china = [True, True, False, True]                # 布尔列表

# 2. 混合类型的列表（合法但不推荐）
mixed = ["杭州", 30.27, 120.16, True]

# 3. 空列表（之后再添加元素）
empty = []

# 打印
print(elevations)
print(cities)
print(empty)
````

### 1.2 列表的索引（与字符串一致）

**列表的索引规则跟字符串完全一样**——你已经会了：

- 从 0 开始
- 支持负索引（从右数，最后一个是 -1）

````python
cities = ["杭州", "北京", "上海", "广州", "深圳"]
#         0      1      2      3      4    ← 正索引
#        -5     -4     -3     -2     -1   ← 负索引

print(cities[0])     # 杭州（第一个）
print(cities[2])     # 上海
print(cities[-1])    # 深圳（最后一个）
print(cities[-2])    # 广州（倒数第二）
````

### 1.3 列表的切片（也跟字符串一样）

````python
cities = ["杭州", "北京", "上海", "广州", "深圳"]

print(cities[0:3])     # ['杭州', '北京', '上海'] —— 前3个（左闭右开！）
print(cities[2:])      # ['上海', '广州', '深圳'] —— 从位置2到末尾
print(cities[:3])      # ['杭州', '北京', '上海'] —— 从开头到位置3（不含）
print(cities[-3:])     # ['上海', '广州', '深圳'] —— 最后3个
print(cities[::-1])    # 反转：['深圳', '广州', '上海', '北京', '杭州']
````

**💡 注意**：列表切片返回的还是**列表**（不是单个元素）——即使切出 1 个元素也是列表。

````python
cities = ["杭州", "北京", "上海"]

# 索引：返回单个元素
x = cities[0]
print(x, type(x))     # 杭州 <class 'str'>

# 切片：返回列表（即使只有一个元素）
y = cities[0:1]
print(y, type(y))     # ['杭州'] <class 'list'>
````

### 1.4 列表的长度：len()

和字符串一样，用 `len()` 获取列表长度（即元素个数）：

````python
cities = ["杭州", "北京", "上海", "广州", "深圳"]
print(f"城市数量：{len(cities)}")    # 5

elevations = []
print(f"空列表长度：{len(elevations)}")    # 0
````

## 二、列表的修改：增、删、改

**列表是'可变'的**——这是它和字符串的最大区别。字符串你不能改单个字符，但列表可以。

### 2.1 修改某个元素

**直接给某个位置赋新值即可**：

````python
cities = ["杭州", "北京", "上海"]

cities[1] = "广州"      # 把第2个元素从'北京'改成'广州'
print(cities)            # ['杭州', '广州', '上海']

# 字符串就不能这么干 —— 字符串不可变
# s = "hello"
# s[0] = "H"    # ❌ 报错！字符串不能改
````

### 2.2 添加元素

**两个最常用的方法**：

| 方法 | 作用 | 例子 |
|---|---|---|
| `append(x)` | 在末尾添加一个元素 | `cities.append("深圳")` |
| `insert(i, x)` | 在位置 i 插入元素 x | `cities.insert(0, "北京")` |

````python
# 1. append —— 末尾添加（最常用！）
cities = ["杭州", "北京"]
cities.append("上海")
print(cities)    # ['杭州', '北京', '上海']

cities.append("广州")
print(cities)    # ['杭州', '北京', '上海', '广州']
````

````python
# 2. insert —— 在指定位置插入
cities = ["杭州", "上海"]
cities.insert(1, "北京")     # 在位置 1 插入'北京'
print(cities)                 # ['杭州', '北京', '上海']

cities.insert(0, "南京")     # 在最前面插入
print(cities)                 # ['南京', '杭州', '北京', '上海']
````

**💡 经典模式：用 append 在循环中收集结果**

````python
# 一个非常常用的模式：用空列表 + append 收集结果
elevations = [2530, 1820, 4200, 3650, 5100, 980, 6200]

high_elevations = []      # 1. 创建空列表

for e in elevations:
    if e > 4000:
        high_elevations.append(e)    # 2. 满足条件就 append

print(f"高海拔点：{high_elevations}")
````

**🌟 这个模式将来会用一千次**——空列表 + 循环 + 条件 + append。

比如：
- 从一堆文件名中筛选出 .tif 文件
- 从一组城市中筛选出在中国的
- 从遥感像元中收集所有植被像元的位置

记住这个**"空列表 + append"模式**——它是数据筛选的基础。

### 2.3 删除元素

**三种常用删除方式**：

| 方法 | 作用 | 例子 |
|---|---|---|
| `remove(x)` | 删除第一个值为 x 的元素 | `cities.remove("北京")` |
| `pop(i)` | 删除位置 i 的元素，并**返回它** | `cities.pop(0)` |
| `del 列表[i]` | 直接删除位置 i 的元素 | `del cities[0]` |

````python
# 1. remove(x) —— 按值删除
cities = ["杭州", "北京", "上海", "广州"]
cities.remove("北京")
print(cities)    # ['杭州', '上海', '广州']

# ⚠️ 如果列表里没有这个值，会报错
# cities.remove("东京")    # ValueError: list.remove(x): x not in list
````

````python
# 2. pop(i) —— 按位置删除并返回
cities = ["杭州", "北京", "上海"]
removed = cities.pop(1)        # 删除位置 1 的元素，并返回它
print(f"删除了：{removed}")     # 删除了：北京
print(cities)                   # ['杭州', '上海']

# pop() 不传参数 —— 删除最后一个
last = cities.pop()
print(f"删除了：{last}")        # 删除了：上海
````

````python
# 3. del —— 直接删除（不返回值）
cities = ["杭州", "北京", "上海"]
del cities[0]
print(cities)    # ['北京', '上海']
````

**💡 三种删除方式的区别**：

- 知道**要删什么值** → `remove(x)`
- 知道**位置**且想**拿到被删的元素** → `pop(i)`
- 知道**位置**且**不需要**被删的元素 → `del 列表[i]`

## 三、列表的查询与排序

### 3.1 查询：in / count / index

| 方法 | 作用 | 例子 |
|---|---|---|
| `x in 列表` | 判断 x 是否在列表中 | `"北京" in cities` → True |
| `count(x)` | 统计 x 出现的次数 | `cities.count("北京")` → 1 |
| `index(x)` | 返回 x 第一次出现的位置 | `cities.index("北京")` → 1 |

````python
cities = ["杭州", "北京", "上海", "北京", "广州"]

# 1. in —— 判断是否存在（最常用！）
print("北京" in cities)        # True
print("东京" in cities)        # False

# 2. count —— 出现次数
print(cities.count("北京"))    # 2（北京出现了2次）
print(cities.count("东京"))    # 0

# 3. index —— 第一次出现的位置
print(cities.index("北京"))    # 1（第一个'北京'在位置1）
# print(cities.index("东京"))  # ❌ ValueError（不存在则报错）
````

**💡 `in` 是最常用的——比 `index()` 更安全**：

```python
# ✅ 安全的做法：先用 in 判断
if "东京" in cities:
    pos = cities.index("东京")
    print(pos)

# ❌ 不安全：直接 index() —— 不存在就报错
# pos = cities.index("东京")    # ValueError
```

### 3.2 排序：sort 和 sorted

**两种排序方式**——理解它们的区别非常重要：

| 方法 | 行为 | 修改原列表？ | 返回值 |
|---|---|---|---|
| `列表.sort()` | 就地排序 | **是** | None |
| `sorted(列表)` | 返回新列表 | 否 | 新的排序后的列表 |

````python
# 1. sort()：就地排序，修改原列表
elevations = [3500, 1200, 6800, 2400, 5500]
elevations.sort()
print(elevations)    # [1200, 2400, 3500, 5500, 6800] —— 原列表被改了

# 倒序
elevations.sort(reverse=True)
print(elevations)    # [6800, 5500, 3500, 2400, 1200]
````

````python
# 2. sorted()：返回新列表，原列表不变
elevations = [3500, 1200, 6800, 2400, 5500]
sorted_e = sorted(elevations)

print("原列表：", elevations)        # [3500, 1200, 6800, 2400, 5500] —— 没变
print("新列表：", sorted_e)          # [1200, 2400, 3500, 5500, 6800]

# 倒序
sorted_desc = sorted(elevations, reverse=True)
print("倒序：", sorted_desc)         # [6800, 5500, 3500, 2400, 1200]
````

**💡 sort vs sorted —— 怎么选？**

- 想要**保留原列表**（很多场景需要） → 用 `sorted()`
- 不在乎原列表 → 用 `sort()`
- **新手默认用 `sorted()` 更安全**——避免不小心修改了原数据

### 3.3 反转：reverse 和切片 [::-1]

````python
cities = ["杭州", "北京", "上海"]

# 方法1：reverse() —— 就地反转
cities.reverse()
print(cities)    # ['上海', '北京', '杭州']

# 方法2：切片 [::-1] —— 返回新列表（原列表不变）
cities = ["杭州", "北京", "上海"]
reversed_cities = cities[::-1]
print("原列表：", cities)              # ['杭州', '北京', '上海']
print("反转后：", reversed_cities)      # ['上海', '北京', '杭州']
````

## 四、元组 —— 不可变的列表

### 4.1 什么是元组？

**元组（tuple）**：和列表很像，但**不可改变**（一旦创建，不能添加、删除、修改元素）。

**特点**：
- 用**圆括号** `()` 创建（注意不是方括号）
- 元素之间用逗号 `,` 分隔
- **创建后不能改**

````python
# 列表 vs 元组
cities_list = ["杭州", "北京", "上海"]      # 列表，方括号
cities_tuple = ("杭州", "北京", "上海")      # 元组，圆括号

print(cities_list, type(cities_list))    # <class 'list'>
print(cities_tuple, type(cities_tuple))  # <class 'tuple'>

# 访问元素的方式跟列表完全一样
print(cities_tuple[0])     # 杭州
print(cities_tuple[-1])    # 上海
print(cities_tuple[0:2])   # ('杭州', '北京')
````

````python
# 元组不能修改
cities_tuple = ("杭州", "北京", "上海")

# cities_tuple[0] = "南京"        # ❌ TypeError: 'tuple' object does not support item assignment
# cities_tuple.append("广州")     # ❌ AttributeError: 'tuple' object has no attribute 'append'
````

### 4.2 元组的'经典场景'：表示坐标对

**元组最常用的场景之一**：表示一组**永远应该一起出现**的数据——比如经纬度坐标。

````python
# 一个城市的坐标（经度、纬度）
beijing = (116.40, 39.90)
shanghai = (121.47, 31.23)
guangzhou = (113.27, 23.13)

# 取出经度和纬度
print(f"北京经度：{beijing[0]}, 纬度：{beijing[1]}")

# 多个城市的坐标可以放进列表（列表里装元组）
cities_coords = [beijing, shanghai, guangzhou]

# 用循环遍历
for coord in cities_coords:
    print(f"经度: {coord[0]}, 纬度: {coord[1]}")
````

**为什么坐标对要用元组而不是列表？**

- 经纬度是一个**整体**——不应该被随意修改、增减
- 元组的'不可变性'就是一种**保护**
- 这是 Python 程序员的**习惯**：表示固定结构用元组，表示动态集合用列表

### 4.3 元组解包 —— 一次取出多个值

**元组的特殊技能**：可以一次性把多个元素取到不同的变量里。

````python
# 元组解包
beijing = (116.40, 39.90)
lon, lat = beijing                    # ★ 一次解包成两个变量
print(f"经度：{lon}, 纬度：{lat}")

# 实用：循环里的解包
cities_coords = [
    ("北京", 116.40, 39.90),
    ("上海", 121.47, 31.23),
    ("广州", 113.27, 23.13),
]

for name, lon, lat in cities_coords:
    print(f"{name}：经度 {lon}, 纬度 {lat}")
````

**🌟 元组解包是 Python 最优雅的特性之一**——以后你会反复用到。

**记住**：列表里装元组、再用 `for x, y, z in 列表` 解包——这是处理'多字段数据'的标准模式。

### 4.4 列表 vs 元组 —— 怎么选？

| 场景 | 用列表 | 用元组 |
|---|---|---|
| 多个观测站的名字（会动态增减） | ✅ | |
| 一个城市的(经度, 纬度) | | ✅ |
| 一组要排序的海拔数据 | ✅ | |
| 一个RGB颜色值(255, 0, 0) | | ✅ |
| 一份会修改的购物清单 | ✅ | |
| 函数返回多个值（下次课会学）| | ✅ |

**经验法则**：
- **会变的数据用列表**
- **固定结构、组合关系紧密的用元组**

## 五、二维列表初探

**列表的元素可以是任何类型——包括另一个列表**。

这就形成了**二维列表**——非常适合表示表格、矩阵、网格。

### 5.1 一个二维列表（迷你 DEM）

````python
# 一个 3行 × 4列 的高程矩阵（迷你 DEM）
dem = [
    [100, 110, 120, 130],     # 第 0 行
    [200, 210, 220, 230],     # 第 1 行
    [300, 310, 320, 330],     # 第 2 行
]

# 整个二维列表
print(dem)

# 第 0 行（一个一维列表）
print(dem[0])    # [100, 110, 120, 130]

# 第 0 行第 2 列的元素 —— 用两个索引！
print(dem[0][2])    # 120
````

**🔑 关键语法**：`矩阵[行][列]` —— 两次索引

- `dem[行][列]` 表示第 `行` 行、第 `列` 列的元素
- 都是从 0 开始数

### 5.2 用嵌套循环遍历二维列表（呼应第5次课）

````python
dem = [
    [100, 110, 120, 130],
    [200, 210, 220, 230],
    [300, 310, 320, 330],
]

rows = len(dem)         # 3
cols = len(dem[0])      # 4（取第0行的长度即列数）

# 嵌套循环遍历每个像元（和第5次课的逻辑一致）
for row in range(rows):
    for col in range(cols):
        print(f"{dem[row][col]:4d}", end=" ")
    print()    # 每行换行
````

````python
# 也可以直接对二维列表遍历（更简洁）
dem = [
    [100, 110, 120, 130],
    [200, 210, 220, 230],
    [300, 310, 320, 330],
]

for row_data in dem:           # row_data 是一个一维列表
    for elevation in row_data:  # 再遍历这个一维列表
        print(f"{elevation:4d}", end=" ")
    print()
````

**💡 二维列表 = NumPy 二维数组的'纯 Python 版'**

第13次课你会看到 NumPy 的二维数组——它和今天的二维列表**结构一样**，但：
- 性能更高（C 语言底层）
- 操作更简洁（不用嵌套循环就能算）
- 专为数值计算优化

**今天的二维列表，是理解 NumPy 数组的根基。**

````python
# ===== 即时练习② 参考答案 =====
stations = ["敦煌", "格尔木", "林芝"]
print("初始：", stations)

# 1. 末尾添加'昆明'
stations.append("昆明")
print("添加昆明后：", stations)

# 2. 最前面插入'塔克拉玛干'
stations.insert(0, "塔克拉玛干")
print("插入塔克拉玛干后：", stations)

# 3. 在'格尔木'之后插入'拉萨'
geermu_pos = stations.index("格尔木")    # 找到'格尔木'位置
stations.insert(geermu_pos + 1, "拉萨")    # 在它后面插入
print("插入拉萨后：", stations)

# 4. 把'林芝'改成'墨脱'
linzhi_pos = stations.index("林芝")
stations[linzhi_pos] = "墨脱"
print("林芝改墨脱后：", stations)

# 5. 删除'昆明'
stations.remove("昆明")
print("删除昆明后：", stations)

# 6. 最终长度
print(f"最终长度：{len(stations)}")
````

````python
# ===== 即时练习③ 任务1 参考答案 =====
stations = [
    ("塔克拉玛干", 1200),
    ("敦煌", 1100),
    ("格尔木", 2800),
    ("拉萨", 3650),
    ("林芝", 2900),
    ("昆明", 1900),
    ("广州", 50),
]

for name, elevation in stations:    # 元组解包
    print(f"观测点：{name}，海拔：{elevation} 米")
````

````python
# ===== 即时练习③ 任务2 参考答案 =====
high_stations = []

for name, elevation in stations:
    if elevation > 2000:
        high_stations.append((name, elevation))

print("海拔 > 2000 米的观测点：")
for name, elevation in high_stations:
    print(f"  {name}: {elevation} 米")
````

````python
# ===== 即时练习③ 任务3 参考答案 =====
max_elevation = 0
max_name = ""

for name, elevation in stations:
    if elevation > max_elevation:
        max_elevation = elevation
        max_name = name

print(f"海拔最高：{max_name}（{max_elevation} 米）")
````

````python
# ===== 即时练习③ 任务4 参考答案 =====
# sorted 默认按元组第一个元素排——也就是按名字（字符串）排
result = sorted(stations)
for s in result:
    print(s)

# 你会发现结果是按名字字典序排，不是按海拔！
# 这就引出了'按指定字段排序'的需求 ——
# 需要用到下次课的'函数'（lambda 或 key 参数）
# 这里给一个超前预告（不强求理解）：
result_by_elevation = sorted(stations, key=lambda x: x[1])
print("\n按海拔排序（超前预告）：")
for s in result_by_elevation:
    print(s)
````

````python
# ===== 课后作业 2：文件名过滤 =====
files = [
    "LC08_123032_20230115.tif",
    "LC09_123032_20230220.shp",
    "LC08_124033_20230310.tif",
    "LC09_125034_20230401.tif",
    "LC08_126035_20230510.shp",
]

# 1. .tif 文件
tif_files = []
for f in files:
    if f.endswith(".tif"):
        tif_files.append(f)

# 2. LC09 卫星
lc09_files = []
for f in files:
    if "LC09" in f:
        lc09_files.append(f)

# 3. 输出
print("=== .tif 文件 ===")
for f in tif_files:
    print(f)
print(f".tif 共 {len(tif_files)} 个")

print("\n=== LC09 文件 ===")
for f in lc09_files:
    print(f)
print(f"LC09 共 {len(lc09_files)} 个")
````

````python
# ===== 课后作业 3：二维 NDVI 网格分析 =====
ndvi_grid = [
    [0.10, 0.15, 0.30, 0.45, 0.60],
    [0.20, 0.35, 0.50, 0.65, 0.75],
    [0.05, 0.25, 0.45, 0.70, 0.80],
    [0.30, 0.40, 0.55, 0.75, 0.85],
]

rows = len(ndvi_grid)        # 4
cols = len(ndvi_grid[0])     # 5
total = rows * cols          # 20

# 1. 总和与平均
total_sum = 0
for row in range(rows):
    for col in range(cols):
        total_sum = total_sum + ndvi_grid[row][col]
average = total_sum / total
print(f"总和：{total_sum:.2f}")
print(f"平均：{average:.3f}")

# 2. 最大 NDVI 的位置
max_ndvi = -1
max_row = 0
max_col = 0
for row in range(rows):
    for col in range(cols):
        if ndvi_grid[row][col] > max_ndvi:
            max_ndvi = ndvi_grid[row][col]
            max_row = row
            max_col = col
print(f"最大 NDVI：{max_ndvi}，位置 [{max_row}, {max_col}]")

# 3. 茂密森林像元数
dense_count = 0
for row in range(rows):
    for col in range(cols):
        if ndvi_grid[row][col] >= 0.7:
            dense_count = dense_count + 1
print(f"茂密森林像元：{dense_count} / {total} ({dense_count/total*100:.1f}%)")
````
