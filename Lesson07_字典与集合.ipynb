## 一、字典的创建与访问

### 1.1 什么是字典？

**字典（dict）**：一个'**键 → 值**'的映射容器。

**特点**：
- 用大括号 `{}` 创建
- 每个元素是一对：`键: 值`（用冒号分隔）
- 不同的键值对之间用逗号分隔
- **键必须唯一**（重复的键会被覆盖）
- **键必须不可变**（数字、字符串、元组都行；列表不行）
- **值可以是任何类型**（数字、字符串、列表、元组、甚至另一个字典）

````python
# 1. 创建一个字典：城市 → 坐标
city_coords = {
    "北京": (116.40, 39.90),
    "上海": (121.47, 31.23),
    "广州": (113.27, 23.13),
    "杭州": (120.16, 30.27),
}

print(city_coords)
print(type(city_coords))    # <class 'dict'>
````

````python
# 2. 不同类型的字典

# 城市 → 海拔
city_elevation = {"北京": 43.5, "拉萨": 3650, "广州": 21}

# 卫星 → 发射年份
satellites = {"Landsat 8": 2013, "Landsat 9": 2021, "Sentinel-2A": 2015}

# 数字 → 字符串
level_names = {1: "水体", 2: "裸土", 3: "稀疏植被", 4: "中等植被", 5: "茂密森林"}

# 空字典
empty = {}
````

### 1.2 访问字典的值：用键

**🔑 语法**：`字典[键]`

````python
city_coords = {
    "北京": (116.40, 39.90),
    "上海": (121.47, 31.23),
    "广州": (113.27, 23.13),
}

# 直接用键取值
print(city_coords["北京"])     # (116.40, 39.90)
print(city_coords["上海"])     # (121.47, 31.23)

# 取出后可以解包成多个变量
lon, lat = city_coords["北京"]
print(f"北京经度：{lon}, 纬度：{lat}")
````

### 1.3 ⚠️ 高频踩坑：访问不存在的键会报错

````python
city_coords = {"北京": (116.40, 39.90), "上海": (121.47, 31.23)}

# 访问不存在的键 → KeyError！
# print(city_coords["东京"])    # ❌ KeyError: '东京'

# 安全的做法：先用 in 判断
if "东京" in city_coords:
    print(city_coords["东京"])
else:
    print("东京不在字典里")
````

### 1.4 ⭐ 更优雅的'安全访问'：get() 方法

**`get(键)` 在键不存在时返回 None（而不是报错）**：

````python
city_coords = {"北京": (116.40, 39.90), "上海": (121.47, 31.23)}

# get() 默认返回 None
print(city_coords.get("北京"))      # (116.40, 39.90)
print(city_coords.get("东京"))      # None（不会报错）

# get() 还能指定'默认值'
print(city_coords.get("东京", "未知"))    # 未知
print(city_coords.get("东京", (0, 0)))    # (0, 0)
````

**🌟 强烈推荐**：除非你**确定**键存在，否则**优先用 `get()`**——比 `dict[key]` 安全得多。

### 1.5 in 运算符（判断键是否存在）

````python
city_coords = {"北京": (116.40, 39.90), "上海": (121.47, 31.23)}

# in 判断的是'键'是否存在（不是值！）
print("北京" in city_coords)               # True
print("东京" in city_coords)               # False

# ⚠️ 注意：in 默认查的是'键'，不是值
print((116.40, 39.90) in city_coords)     # False（这是值，不是键）
````

### 1.6 字典的长度：len()

````python
city_coords = {"北京": (116.40, 39.90), "上海": (121.47, 31.23), "广州": (113.27, 23.13)}

print(f"字典中有 {len(city_coords)} 个城市")    # 3
````

## 二、字典的修改

### 2.1 添加新键值对 / 修改已有值

**Python 把这两件事统一了**——都是用 `字典[键] = 值`：
- 键已存在 → 修改
- 键不存在 → 添加

````python
elevations = {"北京": 43.5, "上海": 4, "广州": 21}

# 1. 添加（键不存在）
elevations["杭州"] = 41.7
print(elevations)    # {'北京': 43.5, '上海': 4, '广州': 21, '杭州': 41.7}

# 2. 修改（键已存在）
elevations["上海"] = 4.5    # 把上海的值改了
print(elevations)    # 上海的值变成 4.5

# 3. 一次添加多个：用 update()
elevations.update({"成都": 506, "西安": 405})
print(elevations)
````

### 2.2 删除键值对：del / pop

````python
elevations = {"北京": 43.5, "上海": 4, "广州": 21}

# 方法1：del 字典[键]
del elevations["上海"]
print(elevations)    # {'北京': 43.5, '广州': 21}

# 方法2：pop(键) —— 删除并返回值
removed = elevations.pop("广州")
print(f"删除了广州，海拔是 {removed}")
print(elevations)    # {'北京': 43.5}
````

**💡 跟列表一样**：
- `del` 简单粗暴，不返回值
- `pop()` 删除并返回被删的值

**⚠️ 警告**：`del` 和 `pop()` 在键不存在时也会报错——所以最好先 `in` 判断或用 `pop(键, 默认值)` 形式。

## 三、字典的遍历

字典支持三种遍历方式：

| 方法 | 遍历什么 | 例子 |
|---|---|---|
| `for k in d` 或 `for k in d.keys()` | 所有的键 | `for city in elevations:` |
| `for v in d.values()` | 所有的值 | `for e in elevations.values():` |
| `for k, v in d.items()` | 所有的键值对（最常用！） | `for city, e in elevations.items():` |

### 3.1 遍历所有的键（默认）

````python
elevations = {"北京": 43.5, "上海": 4, "拉萨": 3650}

# 默认遍历的是键
for city in elevations:
    print(city)

# 等价于：
# for city in elevations.keys():
#     print(city)
````

### 3.2 遍历所有的值

````python
elevations = {"北京": 43.5, "上海": 4, "拉萨": 3650}

for e in elevations.values():
    print(e)
````

### 3.3 ⭐ 同时遍历键和值：items()（最常用！）

````python
elevations = {"北京": 43.5, "上海": 4, "拉萨": 3650}

# items() 返回每个键值对（元组形式），用解包接收
for city, e in elevations.items():
    print(f"{city}：海拔 {e} 米")
````

**🌟 `for k, v in d.items()` 是字典最常用的遍历方式**——记下来。

对比一下：列表遍历用 `for x in lst`；字典遍历用 `for k, v in d.items()`——两个标准模式。

### 3.4 字典遍历的经典应用：统计 / 筛选

````python
elevations = {
    "北京": 43.5,
    "上海": 4,
    "拉萨": 3650,
    "昆明": 1900,
    "广州": 21,
    "林芝": 2900,
}

# 应用1：统计高海拔（>1000米）城市的数量
high_count = 0
for city, e in elevations.items():
    if e > 1000:
        high_count = high_count + 1
print(f"高海拔城市数量：{high_count}")

# 应用2：筛选出高海拔城市（结果是新字典）
high_cities = {}    # 空字典
for city, e in elevations.items():
    if e > 1000:
        high_cities[city] = e    # 添加到新字典
print(f"高海拔城市：{high_cities}")

# 应用3：找出海拔最高的城市
max_city = ""
max_elevation = 0
for city, e in elevations.items():
    if e > max_elevation:
        max_elevation = e
        max_city = city
print(f"海拔最高：{max_city}（{max_elevation} 米）")
````

## 四、集合 set —— 无重复元素的容器

### 4.1 什么是集合？

**集合（set）**：一个**无序、无重复**元素的容器。

**特点**：
- 用大括号 `{}` 创建（注意：和字典看着像，但里面**不是键值对**！）
- 元素**不能重复**——重复会自动去掉
- 元素**没有顺序**——不能用索引访问
- 元素必须**不可变**（数字、字符串、元组都行；列表不行）

**最大用途**：**去重**和**集合运算**（并、交、差）

````python
# 1. 创建集合
cities = {"北京", "上海", "广州"}
print(cities, type(cities))    # <class 'set'>

# 2. 集合的'去重'特性
messy = {"北京", "上海", "北京", "广州", "上海"}    # 重复值
print(messy)    # 自动去重：{'北京', '上海', '广州'}

# 3. ⚠️ 创建空集合不能用 {}（那是字典）
empty1 = {}     # 这是空字典！
empty2 = set()  # 这才是空集合
print(type(empty1))    # <class 'dict'>
print(type(empty2))    # <class 'set'>
````

### 4.2 集合的最常见用途：去重

**用法**：把列表丢进 `set()`，就自动去重了。

````python
# 实用场景：去除列表中的重复值
stations = ["北京", "上海", "北京", "广州", "上海", "杭州", "北京"]

unique = set(stations)
print(unique)    # 自动去重：{'北京', '上海', '广州', '杭州'}
print(f"共 {len(unique)} 个不同的站点")
````

````python
# 如果还想要列表形式（保留'有序'特性），再转回 list
stations = ["北京", "上海", "北京", "广州", "上海"]
unique_list = list(set(stations))
print(unique_list)    # 列表（但顺序可能改变）
````

**🌟 这是集合最实用的用途**——
**"列表 → set → 列表"** 这一招就能去除任何重复元素。

**地学应用**：合并多份观测点名单后去重、统计研究区涉及的不同省份、找出独特的传感器型号……

### 4.3 集合的操作：add / remove / in

````python
cities = {"北京", "上海"}

# 1. add(x) —— 添加元素
cities.add("广州")
print(cities)    # {'北京', '上海', '广州'}

# 重复添加无效（这就是'去重'特性）
cities.add("北京")
print(cities)    # 还是 {'北京', '上海', '广州'} —— 没变化

# 2. remove(x) —— 删除元素（不存在则报错）
cities.remove("上海")
print(cities)    # {'北京', '广州'}

# discard(x) —— 删除元素（不存在也不报错，更安全）
cities.discard("东京")    # 不在里面，但不会报错

# 3. in —— 判断是否存在
print("北京" in cities)    # True
print("东京" in cities)    # False
````

### 4.4 集合可以遍历，但顺序不保证

````python
cities = {"北京", "上海", "广州"}

for city in cities:
    print(city)

# ⚠️ 注意：每次运行的顺序可能不一样
# 集合是'无序'的，不要依赖元素的输出顺序
````

## 五、集合运算 —— 并、交、差

**这是集合最强大的功能**——直接做'数学集合论'里的运算。

### 5.1 四种基本运算

假设有两个集合：
- A = {1, 2, 3, 4}
- B = {3, 4, 5, 6}

| 运算 | 数学符号 | Python 写法 | 结果 |
|---|---|---|---|
| 并集 | A ∪ B | `A \| B` | {1, 2, 3, 4, 5, 6} |
| 交集 | A ∩ B | `A & B` | {3, 4} |
| 差集 | A - B | `A - B` | {1, 2}（A 里有、B 里没有的）|
| 对称差 | A △ B | `A ^ B` | {1, 2, 5, 6}（不重叠的部分）|

````python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print(A | B)    # 并集 {1, 2, 3, 4, 5, 6}
print(A & B)    # 交集 {3, 4}
print(A - B)    # 差集 {1, 2}
print(A ^ B)    # 对称差 {1, 2, 5, 6}
````

### 5.2 🌟 地学场景：研究区域的重叠分析

````python
# 两个研究项目涉及的城市
project_a = {"北京", "上海", "广州", "杭州", "成都"}
project_b = {"上海", "杭州", "南京", "深圳", "广州"}

# 都涉及的城市（交集）—— 重点研究区
shared = project_a & project_b
print(f"两项目共同涉及：{shared}")

# 所有涉及的城市（并集）—— 总研究范围
all_cities = project_a | project_b
print(f"总研究范围：{all_cities}")

# A 独有的城市（差集）—— 特色研究
only_a = project_a - project_b
print(f"项目A 独有：{only_a}")

# 不重叠部分（对称差）—— 互补研究
non_overlap = project_a ^ project_b
print(f"不重叠部分：{non_overlap}")
````

### 5.3 集合 vs 列表 vs 字典 —— 怎么选？

| 场景 | 用什么 |
|---|---|
| 多个数据，**有顺序、可重复** | **列表** |
| 多个数据，**有名字、可查找** | **字典** |
| 多个数据，**无重复** / 需要做并交差 | **集合** |
| 一组永远在一起的数据（如坐标对）| **元组** |

**经验法则**：
- 默认用**列表**
- 需要'按名字找'用**字典**
- 需要'去重'或'集合运算'用**集合**
- 表示'整体不变'用**元组**

## 六、本节课小结

### 知识点回顾

| 知识点 | 关键语法 |
|---|---|
| 创建字典 | `{键1: 值1, 键2: 值2, ...}` |
| 访问值 | `d[key]`（不存在报错）/ `d.get(key, 默认)`（安全） |
| 添加/修改 | `d[key] = 值` |
| 删除 | `del d[key]` / `d.pop(key)` |
| 遍历 | `for k, v in d.items()` |
| 创建集合 | `{元素1, 元素2}` 或 `set()`（空集合）|
| 集合操作 | `add(x)`、`remove(x)`、`discard(x)`、`x in s` |
| 集合运算 | `\|` 并、`&` 交、`-` 差、`^` 对称差 |
| 去重技巧 | `list(set(原列表))` |

### 重点强调（重要程度从高到低）

1. ⭐⭐⭐ **`d.get(key, 默认值)`** —— 访问字典最安全的方式
2. ⭐⭐⭐ **`for k, v in d.items()`** —— 字典遍历的标准模式
3. ⭐⭐ **`list(set(原列表))`** —— 列表去重的一行神技
4. ⭐⭐ **空集合用 `set()` 不是 `{}`** —— `{}` 是空字典
5. ⭐ **集合运算 `|` `&` `-`** —— 处理多组数据重叠的利器

````python
# ===== 即时练习② 参考答案 =====
city_temps = {
    "哈尔滨": 5.6, "北京": 13.2, "上海": 17.1, "广州": 22.5,
    "昆明": 15.7, "乌鲁木齐": 8.0, "拉萨": 8.6, "海口": 24.5,
}

# 1. 遍历输出
for city, temp in city_temps.items():
    print(f"{city}：{temp}°C")

# 2. 寒冷城市数（< 10°C）
cold_count = 0
for city, temp in city_temps.items():
    if temp < 10:
        cold_count = cold_count + 1
print(f"\n寒冷城市：{cold_count} 个")

# 3. 最热城市
max_city = ""
max_temp = 0
for city, temp in city_temps.items():
    if temp > max_temp:
        max_temp = temp
        max_city = city
print(f"最热：{max_city}（{max_temp}°C）")

# 4. 温暖城市筛选
warm_cities = {}
for city, temp in city_temps.items():
    if 10 <= temp <= 22:
        warm_cities[city] = temp
print(f"温暖城市：{warm_cities}")
````

````python
# ===== 即时练习③ 任务2 参考答案 =====
project_a = ["北京", "上海", "杭州", "成都", "广州"]
project_b = ["杭州", "南京", "上海", "深圳"]
project_c = ["广州", "南京", "福州", "成都"]

# 把列表转成集合
set_a = set(project_a)
set_b = set(project_b)
set_c = set(project_c)

# 1. 总集合（并集）
all_cities = set_a | set_b | set_c
print(f"总共涉及 {len(all_cities)} 个城市：{all_cities}")

# 2. 三项目都涉及（交集）
common = set_a & set_b & set_c
print(f"三项目共同：{common}")

# 3. 只在 project_a 中
only_a = set_a - set_b - set_c
print(f"只在 A：{only_a}")
````

````python
# ===== 即时练习③ 任务3 参考答案 =====
project_a = ["北京", "上海", "杭州", "成都", "广州"]
project_b = ["杭州", "南京", "上海", "深圳"]
project_c = ["广州", "南京", "福州", "成都"]

# 用一个字典统计每个城市出现的次数
count_dict = {}

# 遍历所有项目的所有城市
for city in project_a:
    count_dict[city] = count_dict.get(city, 0) + 1

for city in project_b:
    count_dict[city] = count_dict.get(city, 0) + 1

for city in project_c:
    count_dict[city] = count_dict.get(city, 0) + 1

# 输出
print("=== 每个城市出现的项目数 ===")
for city, count in count_dict.items():
    print(f"{city}：{count} 次")
````

````python
# ===== 课后作业 1：城市气候字典查询 =====
climate = {
    "北京": {"temp": 13.2, "rainfall": 580},
    "上海": {"temp": 17.1, "rainfall": 1200},
    "广州": {"temp": 22.5, "rainfall": 1700},
    "昆明": {"temp": 15.7, "rainfall": 1000},
    "哈尔滨": {"temp": 5.6, "rainfall": 530},
}

# 1. 安全访问
beijing = climate.get("北京")
tokyo = climate.get("东京", "未收录")
print(f"北京：{beijing}")
print(f"东京：{tokyo}")

# 2. 遍历输出
print("\n=== 所有城市气候 ===")
for city, info in climate.items():
    print(f"{city}：温度 {info['temp']}°C，降水 {info['rainfall']} mm")

# 3. 筛选高降水城市
rainy_cities = {}
for city, info in climate.items():
    if info["rainfall"] > 1000:
        rainy_cities[city] = info
print(f"\n年降水 > 1000mm：{list(rainy_cities.keys())}")

# 4. 添加拉萨
climate["拉萨"] = {"temp": 8.6, "rainfall": 442}

# 5. 城市数量
print(f"\n最终共有 {len(climate)} 个城市")
````

````python
# ===== 课后作业 2：文件去重与统计 =====
batch1 = ["LC08_001.tif", "LC09_002.tif", "LC08_003.tif"]
batch2 = ["LC09_002.tif", "LC08_004.tif", "LC09_005.tif"]
batch3 = ["LC08_001.tif", "LC09_006.tif", "LC08_003.tif"]

# 1. 合并去重
all_files = set(batch1) | set(batch2) | set(batch3)
print(f"去重后共 {len(all_files)} 个文件")
for f in all_files:
    print(f"  {f}")

# 2. 三批都出现的文件
common = set(batch1) & set(batch2) & set(batch3)
print(f"\n三批都有的文件：{common if common else '无'}")

# 3. 统计每个文件出现在几个批次
file_count = {}
for f in batch1:
    file_count[f] = file_count.get(f, 0) + 1
for f in batch2:
    file_count[f] = file_count.get(f, 0) + 1
for f in batch3:
    file_count[f] = file_count.get(f, 0) + 1

print("\n=== 每个文件出现次数 ===")
for f, count in file_count.items():
    print(f"{f}：{count} 次")
````

````python
# ===== 课后作业 3：NDVI 像元分类统计 =====
import random
random.seed(42)    # 固定随机种子，便于结果重复

# 生成一个 5×5 的 NDVI 网格（用模拟数据）
ndvi_grid = []
for row in range(5):
    row_data = []
    for col in range(5):
        # 用与 row, col 相关的公式生成 0~1 之间的值
        ndvi = (row + col) / 8
        if ndvi > 1:
            ndvi = 1.0
        row_data.append(ndvi)
    ndvi_grid.append(row_data)

# 用字典统计
class_counts = {}

for row in range(5):
    for col in range(5):
        ndvi = ndvi_grid[row][col]
        
        # 分类
        if ndvi < 0:
            class_name = "水体"
        elif ndvi < 0.2:
            class_name = "裸土/建成区"
        elif ndvi < 0.4:
            class_name = "稀疏植被"
        elif ndvi < 0.7:
            class_name = "中等植被"
        else:
            class_name = "茂密森林"
        
        # 累加到字典
        class_counts[class_name] = class_counts.get(class_name, 0) + 1

# 输出
total = 5 * 5
print(f"=== NDVI 五级分类统计（共 {total} 像元）===")
for class_name, count in class_counts.items():
    print(f"{class_name}：{count} 个 ({count/total*100:.1f}%)")
````
