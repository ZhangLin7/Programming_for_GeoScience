## 一、4个高频疑难点回顾

### 🔥 疑难点 1：缩进 + 冒号（来自第4-5次课）

**症状**：写 if/for/while 时不断遇到 `IndentationError` 或 `SyntaxError`。

**核心要点**：
- 凡是后面跟代码块的语句（if、elif、else、for、while），**末尾必须有冒号** `:`
- 代码块必须**缩进**——推荐 4 个空格
- **同一层缩进必须一致**——不能混用 Tab 和空格

**经典踩坑示例**：

````python
# 🔍 找出下面代码的所有错误（不要运行，先用眼看）

elevation = 8848

# if elevation > 8000
#     print("超高峰")
#       print("很高")

# 答：
# 1. if 后面漏了冒号 :
# 2. 第二个 print 缩进多了2个空格（不一致）

# 正确写法：
if elevation > 8000:
    print("超高峰")
    print("很高")
````

**🛡️ 防御技巧**：

1. 写 `if/for/while` 时，**先写冒号 + 回车**——让 Jupyter 自动缩进
2. 看到报错优先检查这两个位置
3. 缩进**统一用 4 个空格**——别按 Tab

### 🔥 疑难点 2：sort vs sorted（来自第6次课）

**症状**：`result = lst.sort()` 然后 `print(result)` —— 出来 `None` 而不是排序后的列表。

**核心要点**：

| 方法 | 行为 | 修改原列表？ | 返回值 |
|---|---|---|---|
| `lst.sort()` | 就地排序 | **是** | None（注意！） |
| `sorted(lst)` | 返回新列表 | 否 | 排序后的新列表 |

````python
# 演示：体会两者区别
elevations = [3500, 1200, 6800, 2400, 5500]

# ❌ 常见错误写法
result = elevations.sort()
print("result =", result)            # None！sort() 不返回排序后的列表
print("原列表 =", elevations)        # 但原列表已经排好序了
````

````python
# ✅ 正确做法（推荐用 sorted）
elevations = [3500, 1200, 6800, 2400, 5500]

sorted_e = sorted(elevations)        # 返回新列表
print("新列表 =", sorted_e)
print("原列表 =", elevations)        # 原列表不变
````

**🛡️ 防御技巧**：

**新手默认用 `sorted()`** —— 更直观、不会改原数据。

### 🔥 疑难点 3：字典 KeyError + get()（来自第7次课）

**症状**：访问字典里不存在的键 → `KeyError`。

**核心要点**：
- `d[key]` —— 键不存在直接报错
- `d.get(key, 默认值)` —— 键不存在返回默认值
- `key in d` —— 先判断再访问

````python
elevations = {"北京": 43.5, "上海": 4, "拉萨": 3650}

# ❌ 不安全（如果东京不存在会崩溃）
# print(elevations["东京"])    # KeyError

# ✅ 安全方法 1：get() + 默认值
print(elevations.get("东京", -1))    # -1

# ✅ 安全方法 2：先 in 判断
if "东京" in elevations:
    print(elevations["东京"])
else:
    print("东京未收录")
````

**🛡️ 防御技巧**：

**默认用 `get()` 或先 `in` 判断**——除非你 100% 确定键存在。

### 🔥 疑难点 4：循环+if 的两大模式（来自第5-7次课）

**这两个模式将来会用一千次**——必须刻进肌肉记忆。

#### 模式 A：累加器（统计数量/总和）

**步骤**：
1. 循环外：初始化变量为 0
2. 循环内：满足条件就累加

````python
# 累加器模式：统计高海拔点（>4000米）
elevations = [2530, 1820, 4200, 3650, 5100, 980, 6200]

count = 0                       # 1. 初始化
for e in elevations:
    if e > 4000:
        count = count + 1       # 2. 累加

print(f"高海拔点：{count} 个")
````

#### 模式 B：空容器 + append（筛选/收集）

**步骤**：
1. 循环外：创建空列表（或空字典）
2. 循环内：满足条件就 append（或字典赋值）

````python
# 列表版：从一组海拔中筛选高海拔点
elevations = [2530, 1820, 4200, 3650, 5100, 980, 6200]

high = []                       # 1. 空列表
for e in elevations:
    if e > 4000:
        high.append(e)          # 2. append

print(f"高海拔点：{high}")

# 字典版：从一组城市中筛选高海拔城市
elevations_dict = {"北京": 43, "拉萨": 3650, "林芝": 2900, "广州": 21}

high_cities = {}                # 1. 空字典
for city, e in elevations_dict.items():
    if e > 1000:
        high_cities[city] = e   # 2. 字典赋值

print(f"高海拔城市：{high_cities}")
````

**🛡️ 这两个模式背下来——是模块一最有价值的肌肉记忆**。

# 📎 附录：综合大练习参考答案

> **强烈建议先独立完成所有任务再看答案。**
> 
> 这是模块一的'集大成'练习，自己写一遍才能真正发现自己的弱点。

````python
# ===== 综合大练习：完整参考答案 =====
raw_data = [
    "BJ001,北京,2024-07-01, 32.5",
    "BJ002,北京,2024-07-02 ,33.8",
    "SH001,上海,2024-07-01,35.2",
    "SH002,上海,2024-07-02,34.1 ",
    "GZ001,广州,2024-07-01,33.5",
    "GZ002,广州,2024-07-02,34.7",
    "KM001,昆明,2024-07-01,22.3",
    "KM002,昆明,2024-07-02,23.5",
    "HRB01,哈尔滨,2024-07-01,28.4",
    "HRB02,哈尔滨,2024-07-02,29.1",
]
````

````python
# ===== 任务1+2：解析字符串，转成 list of tuple =====
records = []

for line in raw_data:
    parts = line.split(",")
    
    # 每个字段用 strip 清理空白
    station_id = parts[0].strip()
    city = parts[1].strip()
    date = parts[2].strip()
    temperature = float(parts[3].strip())    # 温度转 float
    
    # 用元组表示一条记录
    record = (station_id, city, date, temperature)
    records.append(record)

# 验证
for r in records:
    print(r)
````

````python
# ===== 任务3：用集合统计涉及哪些不同的城市 =====
cities = set()    # 空集合

for record in records:
    # record 是 (id, city, date, temp)，city 是第二个
    cities.add(record[1])

print(f"涉及 {len(cities)} 个城市：{cities}")

# 进阶写法：用元组解包
cities2 = set()
for sid, city, date, temp in records:
    cities2.add(city)
print(f"用解包的写法：{cities2}")
````

````python
# ===== 任务4：每个城市的'高温天数'（>30°C）=====
hot_days = {}

for sid, city, date, temp in records:
    if temp > 30:
        # 字典累加模式
        hot_days[city] = hot_days.get(city, 0) + 1

print("=== 各城市高温天数（>30°C）===")
for city, count in hot_days.items():
    print(f"{city}：{count} 天")
````

````python
# ===== 任务5：温度最高的那条记录的观测点 ID =====
max_temp = 0
max_id = ""
max_city = ""

for sid, city, date, temp in records:
    if temp > max_temp:
        max_temp = temp
        max_id = sid
        max_city = city

print(f"最高温度：{max_temp}°C")
print(f"观测点 ID：{max_id}")
print(f"城市：{max_city}")
````

````python
# ===== 任务6：每个城市归并温度列表 + 计算平均温度 =====
city_temps = {}

# 第一步：归并温度到字典
for sid, city, date, temp in records:
    # 如果城市还没在字典里，先创建一个空列表
    if city not in city_temps:
        city_temps[city] = []
    
    # 把这条温度添加到该城市的列表中
    city_temps[city].append(temp)

# 第二步：遍历字典，计算每个城市的平均温度
print("=== 各城市平均温度 ===")
for city, temp_list in city_temps.items():
    total = 0
    for t in temp_list:
        total = total + t
    average = total / len(temp_list)
    print(f"{city}：温度列表 {temp_list}，平均 {average:.2f}°C")
````
