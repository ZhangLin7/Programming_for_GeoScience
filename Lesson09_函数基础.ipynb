## 一、什么是函数？

### 1.1 你其实一直在用函数

回想一下，前 8 次课你用过哪些'看起来像函数'的东西？

```python
print("hello")        # print 是函数
len([1, 2, 3])         # len 是函数
sorted([3, 1, 2])      # sorted 是函数
type(123)              # type 是函数
int("100")             # int 是函数
float("3.14")          # float 是函数
max([1, 5, 3])         # max 是函数
min([1, 5, 3])         # min 是函数
sum([1, 2, 3])         # sum 是函数
range(10)              # range 是函数
```

**这些都是 Python 内置的函数**——你**输入'参数'**，它**返回'结果'**。

### 1.2 函数的本质：一个'代码的盒子'

**函数就是一段被命名了的代码**——你给它'输入'（参数）、它给你'输出'（返回值）：

```
    输入（参数）           输出（返回值）
         ↓                      ↑
    ┌─────────────────┐
    │   函数（黑盒子） │
    │   def 函数名():  │
    │       # 处理...  │
    │       return ... │
    └─────────────────┘
```

**你不需要知道盒子里发生了什么**——只要知道'我喂它什么、它给我什么'。

比如：
- `len([1, 2, 3])` —— 你**不需要**知道 len 内部怎么计算长度的，**反正你给它一个列表，它给你长度**。

### 1.3 为什么需要函数？

**4 个好处**（重要程度从高到低）：

1. ⭐⭐⭐ **避免重复**——一段代码写一次，到处可以用
2. ⭐⭐⭐ **代码模块化**——把大问题拆成小问题，每个函数解决一小块
3. ⭐⭐ **容易修改**——修改逻辑只改一处（不像之前要改很多处）
4. ⭐⭐ **代码更可读**——`calculate_average(temps)` 比一段循环代码更直观

**今天起，写代码的'思维模式'要改变**：
- 之前：从上到下写一长段代码
- 现在：思考'这件事可以拆成哪些小函数？'

## 二、定义自己的函数：def

### 2.1 最简单的函数

**🔑 语法**：

```python
def 函数名():
    要执行的代码
```

- `def` 是关键字（define，定义）
- 函数名后面有**括号 `()`**
- 末尾有**冒号 `:`**
- 函数体要**缩进**（和 if/for 一样）

````python
# 定义一个最简单的函数：打印分隔线
def print_separator():
    print("=" * 30)
    print("GIS 数据分析")
    print("=" * 30)

# 调用函数：函数名 + ()
print_separator()
print_separator()    # 想用多少次都行
````

**关键观察**：
- `def` 定义了函数——但**没有立即执行**
- 必须**调用**（`print_separator()`）才会真正执行
- **可以重复调用**（这就是函数的价值）

### 2.2 带参数的函数

**没有参数的函数太死板**——上面那个函数只能打印固定的分隔线。我们让它**接收一个参数**，根据参数变化：

````python
# 函数有了参数，就更灵活了
def print_separator(title):
    print("=" * 30)
    print(title)
    print("=" * 30)

# 调用时传入不同的参数
print_separator("GIS 数据分析")
print()
print_separator("NDVI 计算结果")
print()
print_separator("杭州气候报告")
````

**关键观察**：
- `title` 是**参数**——它在函数内部就像一个'变量'
- 调用时传入的值，会赋给参数
- **参数让函数变灵活**——同一个函数处理不同的输入

### 2.3 带返回值的函数：return

上面的函数只是**打印**——它没有'返回'东西。

**真正有用的函数应该返回一个结果**——用 `return` 关键字：

````python
# 一个会返回结果的函数：摄氏度转华氏度
def celsius_to_fahrenheit(celsius):
    fahrenheit = celsius * 9 / 5 + 32
    return fahrenheit       # 把结果'返回'给调用者

# 调用并接收返回值
result = celsius_to_fahrenheit(25)
print(f"25°C = {result}°F")

# 也可以直接用在表达式中
print(f"100°C = {celsius_to_fahrenheit(100)}°F")
print(f"0°C = {celsius_to_fahrenheit(0)}°F")
````

**关键观察**：
- `return` 返回一个结果给**调用者**
- 调用者用一个变量**接收**这个结果（`result = ...`）
- **遇到 `return` 函数立即结束**——后面的代码不会执行

### 2.4 ⚠️ print 和 return 的区别（高频踩坑）

**这是新手最容易混淆的两个概念**：

| 行为 | print | return |
|---|---|---|
| 作用 | **打印**到屏幕给人看 | **返回**值给调用者用 |
| 谁能看到？ | 人类（屏幕上的文字）| 程序（调用方接收） |
| 函数能继续执行吗？| 能 | 不能（函数立即结束）|

````python
# ❌ 用 print 而不是 return
def bad_to_fahrenheit(celsius):
    fahrenheit = celsius * 9 / 5 + 32
    print(fahrenheit)            # 打印到屏幕，但没有返回

result = bad_to_fahrenheit(25)
print(f"result = {result}")      # result = None ❌
````

````python
# ✅ 用 return
def good_to_fahrenheit(celsius):
    fahrenheit = celsius * 9 / 5 + 32
    return fahrenheit            # 返回结果

result = good_to_fahrenheit(25)
print(f"result = {result}")      # result = 77.0 ✅

# 返回的值还可以参与计算
double = good_to_fahrenheit(25) * 2
print(f"double = {double}")
````

**🛡️ 防御口诀**：

- **想'让用户看到结果'** → 用 `print`
- **想'让别处代码用这个结果'** → 用 `return`
- 两者**经常组合用**——函数 return 一个值，外面 print 显示给用户

## 三、多个参数 + 默认参数

### 3.1 多个参数

函数可以有**任意多个参数**，用逗号分隔：

````python
# 计算两点之间的距离（简化的欧氏距离）
def distance(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    return (dx ** 2 + dy ** 2) ** 0.5    # 平方根 = 0.5 次方

# 调用
d1 = distance(0, 0, 3, 4)
print(f"原点到(3,4)：{d1}")    # 5.0

d2 = distance(116.40, 39.90, 121.47, 31.23)    # 北京到上海（简化）
print(f"北京到上海：{d2:.4f}")
````

**注意**：
- 调用时**参数顺序很重要**——`distance(0, 0, 3, 4)` 和 `distance(3, 4, 0, 0)` 的 `x1, y1, x2, y2` 含义不同（虽然结果一样）
- 参数名是**给函数内部用的**——外面调用时只需要传**值**

### 3.2 默认参数（很实用）

**有时候希望参数有'默认值'**——调用者不传也行，传了就用新的：

````python
# 一个'格式化分隔线'函数：默认 30 个等号
def make_separator(length=30, char="="):
    return char * length

# 不传参数 —— 用默认值
print(make_separator())              # 30 个 =

# 只传第 1 个参数
print(make_separator(20))            # 20 个 =

# 传两个参数
print(make_separator(20, "-"))       # 20 个 -

# 用'关键字参数'明确指定（更可读）
print(make_separator(char="*"))      # 默认 30 个 *
````

**关键观察**：
- 默认参数用 `参数名=默认值` 的语法定义
- 调用时**可以省略**有默认值的参数
- 也可以用'**关键字参数**'明确指定（`char="*"`）—— 更可读

**📌 重要规则**：**有默认值的参数必须放在没有默认值的参数后面**

````python
# ✅ 正确：必填在前，有默认值在后
def good(name, age=18):
    print(f"{name}, {age}")

# ❌ 错误：有默认值的在前，必填在后（语法错误）
# def bad(age=18, name):
#     print(...)    # SyntaxError
````

## 四、返回多个值

**Python 函数可以一次返回多个值**——这是 Python 的特色之一。

### 4.1 用元组返回多个值

````python
# 一次性返回最大值、最小值、平均值
def analyze(elevations):
    max_e = max(elevations)
    min_e = min(elevations)
    avg_e = sum(elevations) / len(elevations)
    return max_e, min_e, avg_e        # 返回 3 个值（实际是一个元组）

# 用 3 个变量接收（元组解包）
data = [3500, 1200, 6800, 2400, 5500]
high, low, mean = analyze(data)
print(f"最大：{high}, 最小：{low}, 平均：{mean:.2f}")
````

**关键观察**：
- `return a, b, c` 实际上**返回一个元组** `(a, b, c)`
- 调用时用**元组解包**接收（学过的，第6次课的元组解包）
- **如果只用一个变量接收**，得到的就是元组本身：

````python
# 用一个变量接收 —— 得到元组
result = analyze([3500, 1200, 6800, 2400, 5500])
print(result)            # (6800, 1200, 3880.0)
print(type(result))      # <class 'tuple'>

# 之后用索引取出
print(result[0])         # 6800（最大）
print(result[1])         # 1200（最小）
````

### 4.2 一个实用例子：解析坐标字符串

````python
# 把第3次课的'坐标解析'封装成函数
def parse_dms(coord_str):
    """把度分秒字符串解析成 (度, 分, 秒) 三元组"""
    parts = coord_str.replace("°", " ").replace("'", " ").replace('"', " ").split()
    deg = int(parts[0])
    minute = int(parts[1])
    second = int(parts[2])
    return deg, minute, second

# 调用：用解包接收
d, m, s = parse_dms("39°54'27\"")
print(f"度：{d}, 分：{m}, 秒：{s}")
````

**🌟 看到了吗？**——上次课你写过这段解析代码——**今天把它封装成函数 `parse_dms`**——以后任何地方都能调用——**这就是函数的价值**。

## 五、函数的'作用域'（点到为止）

**作用域**：变量能被访问的范围。

### 5.1 函数内部的变量是'局部'的

**最重要的一条**：**函数内部定义的变量，外面看不到**：

````python
def my_function():
    x = 100              # x 是函数内部的'局部变量'
    print(f"函数内 x = {x}")

my_function()

# 在函数外面访问 x ?
# print(x)             # ❌ NameError: name 'x' is not defined
````

**类比**：
- 函数像一个'**抽屉**'——里面的变量是'抽屉里的东西'，从外面看不见
- 你只能通过'**返回值**'把抽屉里的东西拿出来

### 5.2 外面的变量函数内部能不能用？

````python
# 外面的变量，函数内部'读取'是可以的（但不推荐）
PI = 3.14159    # 全局变量（函数外面定义）

def circle_area(radius):
    return PI * radius ** 2    # 这里能读到 PI

print(circle_area(5))    # 78.53975
````

**📌 经验法则**：

- **函数内部的变量** → **不要假设外面能看到**（基本看不到）
- **函数外部的变量** → 函数内部能**读取**（但不推荐过度依赖）
- **最佳实践**：函数所需的所有数据都通过**参数**传入；函数的所有结果都通过 **return** 传出

**💡 这样写出来的函数最清晰、最可重用——这就是'函数式思维'。**

### 5.3 ⚠️ 同名变量不会互相影响

````python
x = 100    # 外面的 x

def func():
    x = 999    # 函数内部的 x（和外面的 x 是不同的两个变量）
    print(f"函数内 x = {x}")    # 999

func()
print(f"函数外 x = {x}")    # 100（外面的 x 没变！）
````

**结论**：函数内部对变量的修改，**不会影响外面的同名变量**。这是 Python 设计上保护的——**函数不会'污染'外部环境**。

（注：这条规则有个例外是修改可变对象，比如列表，**今天先不展开**——等你写多了自然就懂。）

````python
# ===== 即时练习③ 任务2 参考答案 =====
def count_hot_days(records, threshold=30):
    """统计每个城市超过阈值的天数"""
    result = {}
    for sid, city, date, temp in records:
        if temp > threshold:
            result[city] = result.get(city, 0) + 1
    return result

# 简单测试
test_records = [
    ("A", "北京", "2024-07-01", 32.0),
    ("B", "北京", "2024-07-02", 28.0),
    ("C", "上海", "2024-07-01", 35.0),
]
print(count_hot_days(test_records))    # 应输出 {'北京': 1, '上海': 1}
````

## 课后作业参考答案

````python
# ===== 课后作业 1：3 个简单函数 =====

# 1. 圆面积
def area_circle(r):
    PI = 3.14159
    return PI * r ** 2

print(area_circle(5))     # 78.53975
print(area_circle(10))    # 314.159

# 2. 温度等级
def temperature_level(temp):
    if temp < 0:
        return "严寒"
    elif temp < 10:
        return "寒冷"
    elif temp < 20:
        return "凉爽"
    elif temp < 30:
        return "温暖"
    else:
        return "炎热"

print(temperature_level(-5))    # 严寒
print(temperature_level(15))    # 凉爽
print(temperature_level(35))    # 炎热

# 3. 度分秒转十进制（含方向）
def dms_to_decimal(degrees, minutes, seconds, direction):
    decimal = degrees + minutes / 60 + seconds / 3600
    if direction == "S" or direction == "W":
        decimal = -decimal
    return decimal

print(dms_to_decimal(39, 54, 27, "N"))    # 39.9075
print(dms_to_decimal(39, 54, 27, "S"))    # -39.9075
print(dms_to_decimal(116, 24, 27, "E"))   # 116.4075
````

````python
# ===== 课后作业 2：NDVI 分类函数 =====

def classify_ndvi(ndvi):
    if ndvi < 0:
        return "水体"
    elif ndvi < 0.2:
        return "裸土/建成区"
    elif ndvi < 0.4:
        return "稀疏植被"
    elif ndvi < 0.7:
        return "中等植被"
    else:
        return "茂密森林"

# 测试 5 个值
test_values = [-0.1, 0.1, 0.3, 0.5, 0.8]
for v in test_values:
    print(f"NDVI = {v} → {classify_ndvi(v)}")
````

````python
# ===== 课后作业 3：列表分析函数集 =====

def count_above(numbers, threshold):
    """统计大于阈值的元素个数"""
    count = 0
    for n in numbers:
        if n > threshold:
            count = count + 1
    return count

def filter_above(numbers, threshold):
    """返回大于阈值的元素的新列表"""
    result = []
    for n in numbers:
        if n > threshold:
            result.append(n)
    return result

def summary(numbers):
    """返回最大、最小、平均、个数"""
    return max(numbers), min(numbers), sum(numbers)/len(numbers), len(numbers)

# 测试
elevations = [2530, 1820, 4200, 3650, 5100, 980, 6200]

print(f"高于 4000 米的点数：{count_above(elevations, 4000)}")
print(f"高于 4000 米的列表：{filter_above(elevations, 4000)}")

high, low, avg, n = summary(elevations)
print(f"最高 {high}, 最低 {low}, 平均 {avg:.1f}, 共 {n} 个")
````
