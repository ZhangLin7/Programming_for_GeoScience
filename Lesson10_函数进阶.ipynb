## 一、lambda 匿名函数

### 1.1 什么是 lambda？

**lambda 是一种'轻量级函数'**：
- 没有名字（**匿名**）
- 只有一行代码
- 用来在'**短暂、临时'的场景下快速定义函数

**🔑 语法**：

```python
lambda 参数: 表达式
```

**对比 def**：

````python
# 普通 def 写法
def double(x):
    return x * 2

print(double(5))    # 10

# lambda 写法（一行搞定）
double_lambda = lambda x: x * 2
print(double_lambda(5))    # 10
````

**关键观察**：
- `lambda x: x * 2` 是一个'**函数对象**'——可以赋值给变量，也可以直接当参数传
- 没有 `def` 关键字、没有 `return` 关键字（冒号后面就是返回值）
- 只能写**一个表达式**——不能有多行、不能有 if/for 复杂逻辑

### 1.2 ⚠️ lambda 的'适用场景'

**lambda 不是用来替代 def 的**——它有特定用武之地：

**✅ 适合 lambda 的场景**：
- 函数很短（一行就够）
- 函数只用一次
- 作为 `sorted`/`max`/`min`/`map`/`filter` 的参数

**❌ 不适合 lambda 的场景**：
- 函数复杂（多行逻辑）
- 函数会被多次调用
- 函数需要文档说明

**经验法则**：复杂的写 def，简单的写 lambda。

### 1.3 ⭐ lambda 最经典场景：sorted 的 key 参数

**这就是第6次课的遗留问题**——怎么按指定字段排序？

````python
stations = [
    ("塔克拉玛干", 1200),
    ("敦煌", 1100),
    ("拉萨", 3650),
    ("林芝", 2900),
    ("昆明", 1900),
    ("广州", 50),
]

# ❌ 默认排序：按元组的第一个元素（城市名）—— 字母顺序
result1 = sorted(stations)
for s in result1:
    print(s)
````

````python
# ✅ 用 lambda 告诉 sorted '按海拔排序'
result2 = sorted(stations, key=lambda x: x[1])

print("按海拔升序：")
for s in result2:
    print(s)
````

**🌟 解读 `key=lambda x: x[1]`**：
- `key=` 告诉 sorted '比较时用什么作为依据'
- `lambda x: x[1]` 是一个匿名函数：**接收一个元素 x，返回它的第二项**（也就是海拔）
- sorted 内部会对每个元素调用这个 lambda，得到一个'比较值'，然后按它排序

````python
# 倒序：海拔从高到低
result3 = sorted(stations, key=lambda x: x[1], reverse=True)

print("按海拔降序：")
for s in result3:
    print(s)
````

### 1.4 🛠️ 更多 lambda + sorted 的例子

````python
# 例1：按城市名长度排序
cities = ["杭州", "哈尔滨", "上海", "乌鲁木齐", "北京"]

by_length = sorted(cities, key=lambda x: len(x))
print(by_length)    # 按长度升序
````

````python
# 例2：用字典排序——按值排序
elevations = {"北京": 43.5, "拉萨": 3650, "昆明": 1900, "广州": 21}

# d.items() 返回 (key, value) 元组的序列
# 按 value（海拔）排序
sorted_items = sorted(elevations.items(), key=lambda x: x[1])

for city, e in sorted_items:
    print(f"{city}：{e} 米")
````

````python
# 例3：复杂数据 —— 城市坐标 (name, lon, lat)
cities = [
    ("北京", 116.40, 39.90),
    ("上海", 121.47, 31.23),
    ("广州", 113.27, 23.13),
    ("哈尔滨", 126.65, 45.75),
    ("昆明", 102.83, 25.04),
]

# 按纬度（第3项，索引2）排序
by_lat = sorted(cities, key=lambda x: x[2])
for c in by_lat:
    print(c)
````

### 1.5 lambda 配合 max / min

````python
stations = [
    ("塔克拉玛干", 1200),
    ("敦煌", 1100),
    ("拉萨", 3650),
    ("林芝", 2900),
]

# 找海拔最高的站
highest = max(stations, key=lambda x: x[1])
print(f"海拔最高：{highest}")

# 找海拔最低的站
lowest = min(stations, key=lambda x: x[1])
print(f"海拔最低：{lowest}")

# 注意：和上次课你写的'最大值查找循环'相比，这一行更简洁
````

**这就是 lambda 的力量**——让代码更简洁、更专业。

## 二、模块导入 —— 用 Python 强大的标准库

**Python 自带海量的'工具库'**——不用我们自己写。**用之前先 `import` 导入**。

### 2.1 三种导入方式

**🔑 方式一**：`import 模块`（推荐）

````python
import math    # 导入数学库

# 用法：模块名.函数名
print(math.pi)              # 3.141592653589793
print(math.sqrt(16))        # 4.0（开平方）
print(math.sin(math.pi/2))  # 1.0（正弦）
print(math.log(math.e))     # 1.0（自然对数）
````

**🔑 方式二**：`from 模块 import 函数`（明确指定要用什么）

````python
from math import pi, sqrt    # 只导入 pi 和 sqrt

# 用法：直接用，不需要 math.
print(pi)                # 3.141592653589793
print(sqrt(16))          # 4.0

# math.cos 还能用吗？不能，因为没导入
# print(cos(0))          # NameError
````

**🔑 方式三**：`import 模块 as 别名`（给模块起个短名字）

````python
import math as m    # 把 math 起个短名字 m

print(m.pi)
print(m.sqrt(25))

# 这种写法在用 numpy/pandas 时非常常见：
# import numpy as np
# import pandas as pd
````

### 2.2 三种方式的选择

| 方式 | 优点 | 缺点 | 适用 |
|---|---|---|---|
| `import math` | 清楚知道函数来自哪个模块 | 写起来略长 | **推荐默认用** |
| `from math import pi` | 直接用，不用前缀 | 和其他变量易冲突 | 函数名很常用且不冲突时 |
| `import math as m` | 简洁 | 别名要约定俗成 | 长名模块（numpy/pandas）|

### 2.3 几个常用的标准库（地学相关）

````python
# 1. math 模块 —— 数学计算
import math

# 常用：sqrt, pi, sin, cos, log, ceil, floor
radius = 5
area = math.pi * radius ** 2
circumference = 2 * math.pi * radius
print(f"半径 {radius} 的圆：面积 {area:.2f}, 周长 {circumference:.2f}")

# ceil 向上取整、floor 向下取整
print(math.ceil(3.2))    # 4
print(math.floor(3.8))   # 3
````

````python
# 2. random 模块 —— 随机数（生成测试数据有用）
import random

# 随机浮点数 [0, 1)
print(random.random())

# 随机整数（包含两端）
print(random.randint(1, 100))

# 随机选择列表中的一个元素
cities = ["北京", "上海", "广州", "杭州"]
print(random.choice(cities))

# 模拟 5 个观测点的随机海拔（3000-6000 米）
for i in range(5):
    elevation = random.randint(3000, 6000)
    print(f"观测点 {i+1}：{elevation} 米")
````

````python
# 3. statistics 模块 —— 统计计算
import statistics

data = [25, 32, 28, 35, 30, 27, 33, 29]

print(f"平均值：{statistics.mean(data):.2f}")
print(f"中位数：{statistics.median(data)}")
print(f"标准差：{statistics.stdev(data):.2f}")
````

**💡 体会一下**——你不用自己写'计算标准差'的循环——**统计模块直接给你了**——这就是 import 的力量。

## 三、docstring —— 给函数写说明

### 3.1 什么是 docstring？

**docstring** 是写在**函数体第一行的字符串**——用来说明这个函数干什么。

**用三引号 `"""..."""` 包起来**：

````python
def calculate_bmi(weight, height):
    """计算 BMI 指数。

    参数：
        weight (float): 体重（千克）
        height (float): 身高（米）
    
    返回：
        float: BMI 值
    """
    return weight / (height ** 2)

# 用 help() 可以看 docstring
help(calculate_bmi)
````

**docstring 的好处**：
- 别人（包括 3 个月后的你自己）能快速看懂函数干什么
- IDE 能自动显示函数说明
- `help(函数名)` 直接显示说明
- **专业代码的标准做法**

**📌 经验法则**：**所有'要给别人用的函数'都应该有 docstring**。

### 3.2 docstring 的几种风格

**简单风格**（适合小函数）：

````python
def square(x):
    """返回 x 的平方。"""
    return x ** 2

help(square)
````

**完整风格**（适合复杂函数，参考上面 calculate_bmi 的写法）：

```
"""一句话简介。

更详细的说明（可省略）。

参数：
    参数1 (类型): 说明
    参数2 (类型): 说明

返回：
    类型: 说明
"""
```

**📌 今天起，希望你写的每个函数都至少有一句话 docstring**。

## 四、`*args` 和 `**kwargs` —— 不定参数（点到为止）

**这两个语法看起来奇怪**，但你在以后的代码里**经常会看到**——所以今天**让你能看懂**就行，**不强求会写**。

### 4.1 `*args` —— 接收任意多个位置参数

**问题**：怎么写一个函数能接收 1 个、2 个、3 个、… 任意多个参数？

````python
# 加 * 号让参数能接收任意多个
def average(*args):
    """计算所有参数的平均值。
    
    支持任意多个参数。
    """
    if len(args) == 0:
        return 0
    return sum(args) / len(args)

# 调用时可以传任意多个参数
print(average(10, 20))                # 15.0
print(average(10, 20, 30))            # 20.0
print(average(1, 2, 3, 4, 5, 6, 7))   # 4.0
````

**关键观察**：
- 参数前的 `*` 表示'**收集所有位置参数到一个元组**'
- 函数内部 `args` 是一个**元组**
- `args` 这个名字是**约定俗成**——你也可以叫 `numbers`、`values`，但 `args` 是惯例

**实际上你已经用过了** —— `print("a", "b", "c")` 就是用 `*args` 实现的：print 函数能接收任意多个参数。

### 4.2 `**kwargs` —— 接收任意多个关键字参数

**问题**：怎么写一个函数能接收 `name="Alice", age=25, city="北京"` 这种任意多的'键值对'参数？

````python
# 加 ** 让参数能接收任意多个'关键字参数'
def print_info(**kwargs):
    """打印所有传入的关键字参数。"""
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="北京", lat=39.90, lon=116.40)
print()
print_info(city="拉萨", elevation=3650)
````

**关键观察**：
- 参数前的 `**` 表示'**收集所有关键字参数到一个字典**'
- 函数内部 `kwargs` 是一个**字典**

**📌 今天先了解就行**——你以后看到别人代码里有 `*args`、`**kwargs`，**能看懂就好**。

**典型应用场景**——封装库函数（让你的函数能透明地传递参数给底层函数）。

## 五、函数作为参数 —— 高阶函数入门

**这是 Python 的'高阶'特性**——但你今天上午其实已经用过了：`sorted(stations, key=lambda x: x[1])` —— **这里 `key` 接收的就是一个函数**。

### 5.1 函数也是'值'，可以传递

**Python 中函数和数字、字符串、列表一样，是'对象'**——可以赋值给变量、可以作为参数传递。

````python
def double(x):
    return x * 2

# 函数也是值！可以赋值给变量
f = double         # 注意：没有括号 —— 不是调用，是引用
print(f(5))        # 10（用 f 调用，效果和 double 一样）
````

### 5.2 把函数作为参数传给另一个函数

````python
def double(x):
    return x * 2

def square(x):
    return x ** 2

# 写一个'通用工具'：对列表里每个元素应用某个函数
def apply_to_list(numbers, func):
    """对 numbers 列表里每个元素应用 func 函数，返回新列表。"""
    result = []
    for n in numbers:
        result.append(func(n))
    return result

# 调用：传入不同的函数
data = [1, 2, 3, 4, 5]

doubled = apply_to_list(data, double)        # 用 double
squared = apply_to_list(data, square)        # 用 square
lambda_result = apply_to_list(data, lambda x: x + 100)    # 直接传 lambda

print("原始：", data)
print("翻倍：", doubled)
print("平方：", squared)
print("加100：", lambda_result)
````

**🌟 高阶函数的价值**：
- 你写一次 `apply_to_list` —— 之后**对任何函数都能复用**
- 代码非常**通用**和**模块化**

**这就是 sorted、max、min 的设计哲学**——它们是**通用工具**，让用户传函数告诉它'怎么比较'。

### 5.3 Python 内置的高阶函数：map 和 filter（点到为止）

**map** 和 **filter** 是 Python 内置的高阶函数——**对列表批量处理**。

````python
# map(函数, 列表) —— 对每个元素应用函数
data = [1, 2, 3, 4, 5]

# 把每个元素平方
result = list(map(lambda x: x ** 2, data))
print(result)    # [1, 4, 9, 16, 25]
````

````python
# filter(函数, 列表) —— 筛选满足条件的元素
data = [3, -1, 5, -2, 8, -4, 6]

# 筛选出正数
positives = list(filter(lambda x: x > 0, data))
print(positives)    # [3, 5, 8, 6]
````

**注意**：`map`、`filter` 返回的不是列表，而是'迭代器'——必须用 `list()` 转换才能看到结果。

**📌 经验法则**：**新手优先用循环**——map/filter 等到代码写多了再用。

````python
# ===== 即时练习② 参考答案 =====
import math

def geographic_distance(lat1, lon1, lat2, lon2):
    """用简化公式计算两点的地理距离（公里）。
    
    参数：
        lat1, lon1 (float): 第一个点的纬度和经度
        lat2, lon2 (float): 第二个点的纬度和经度
    
    返回：
        float: 距离（公里）
    """
    # 平均纬度（弧度）
    avg_lat_rad = math.radians((lat1 + lat2) / 2)
    
    # 经纬度差
    dlat = lat1 - lat2
    dlon = (lon1 - lon2) * math.cos(avg_lat_rad)
    
    # 距离（111 公里 = 1 度纬度的近似值）
    return 111 * math.sqrt(dlat ** 2 + dlon ** 2)

# 测试：北京到上海
d = geographic_distance(39.90, 116.40, 31.23, 121.47)
print(f"北京到上海：约 {d:.0f} 公里")
````

````python
# ===== 即时练习③ 任务3 参考答案 =====
def filter_stations(stations, condition_func):
    """按条件函数筛选观测点。
    
    参数：
        stations (list): 观测点列表
        condition_func: 接收一个观测点，返回 True/False 的函数
    
    返回：
        list: 满足条件的观测点列表
    """
    result = []
    for s in stations:
        if condition_func(s):
            result.append(s)
    return result

# 筛选降水量 > 1000
wet = filter_stations(stations, lambda x: x[2] > 1000)
print("=== 降水量 > 1000 ===")
for s in wet:
    print(s)

# 筛选纬度 < 30
south = filter_stations(stations, lambda x: x[1] < 30)
print("\n=== 纬度 < 30 ===")
for s in south:
    print(s)
````

````python
# ===== 课后作业 1：lambda + sorted =====
stations = [
    ("哈尔滨", 45.75, 530),
    ("北京", 39.90, 580),
    ("上海", 31.23, 1200),
    ("昆明", 25.04, 1000),
    ("广州", 23.13, 1700),
    ("拉萨", 29.65, 442),
]

# 1. 按城市名长度
by_name_len = sorted(stations, key=lambda x: len(x[0]))
print("=== 按城市名长度 ===")
for s in by_name_len:
    print(s)

# 2. 按降水量与平均降水量之差的绝对值
import statistics
avg = statistics.mean([s[2] for s in stations])
by_diff = sorted(stations, key=lambda x: abs(x[2] - avg))
print(f"\n=== 按距平均降水量({avg:.0f})的差距 ===")
for s in by_diff:
    print(f"{s} 距平均：{abs(s[2] - avg):.0f}")

# 3. 纬度最高且降水量 > 500（先筛选再 max）
candidates = [s for s in stations if s[2] > 500]
if candidates:
    result = max(candidates, key=lambda x: x[1])
    print(f"\n纬度最高且降水量>500：{result}")
````

````python
# ===== 课后作业 2：用标准库 =====
import random
import statistics

random.seed(42)    # 固定种子，结果可复现

# 1. 生成 100 个 NDVI 模拟值
ndvi_values = []
for i in range(100):
    ndvi_values.append(random.uniform(-0.2, 1.0))

# 2. 统计
print(f"均值：{statistics.mean(ndvi_values):.3f}")
print(f"中位数：{statistics.median(ndvi_values):.3f}")
print(f"标准差：{statistics.stdev(ndvi_values):.3f}")

# 3. 排序后看最大、最小的 5 个
sorted_ndvi = sorted(ndvi_values)
print(f"\n最小 5 个：{[round(v, 3) for v in sorted_ndvi[:5]]}")
print(f"最大 5 个：{[round(v, 3) for v in sorted_ndvi[-5:]]}")

# 4. NDVI > 0.5 的像元数
count = 0
for v in ndvi_values:
    if v > 0.5:
        count = count + 1
print(f"\n茂密植被（NDVI > 0.5）：{count} 个")
````

````python
# ===== 课后作业 3：通用排序工具 =====
stations = [
    ("哈尔滨", 45.75, 530),
    ("北京", 39.90, 580),
    ("上海", 31.23, 1200),
    ("昆明", 25.04, 1000),
    ("广州", 23.13, 1700),
    ("拉萨", 29.65, 442),
]

def find_extreme(stations, key_func, n=3):
    """返回 stations 按 key_func 降序排序后的前 n 名。
    
    参数：
        stations (list): 观测点列表
        key_func: 排序的'键函数'（接收一个观测点，返回比较值）
        n (int): 返回的数量，默认 3
    
    返回：
        list: 前 n 名观测点
    """
    sorted_list = sorted(stations, key=key_func, reverse=True)
    return sorted_list[:n]

# 降水量前 3
top_rain = find_extreme(stations, lambda x: x[2], 3)
print("=== 降水量前 3 ===")
for s in top_rain:
    print(s)

# 纬度后 3（最南的 3 个）
# 用降序的方式取后 3 = 用升序取前 3
south_3 = find_extreme(stations, lambda x: -x[1], 3)    # 加负号实现升序
print("\n=== 纬度最低的 3 个 ===")
for s in south_3:
    print(s)

# 也可以用 help() 看 docstring
help(find_extreme)
````
