## 一、Pandas 的两个核心结构

### 1.1 安装确认 + import

````python
import pandas as pd
import numpy as np

print(pd.__version__)    # 验证已安装
````

### 1.2 Series —— 带'标签'的一维数组

**Series 像 NumPy 的一维数组**——但每个元素**有标签（index）**：

```
  index    value
  -------  -----
  北京     43
  上海     4
  广州     21
  拉萨     3650
```

````python
# 从字典创建 Series
elevations = pd.Series({"北京": 43, "上海": 4, "广州": 21, "拉萨": 3650})
print(elevations)
print("---")

# 用'标签'访问（像字典）
print("拉萨海拔：", elevations["拉萨"])

# Series 也支持向量化运算（继承自 NumPy）
print("---")
print("海拔翻倍：")
print(elevations * 2)
````

**关键观察**：
- Series 有'**index**'（左列）和'**values**'（右列）
- 既能像字典访问 `s["拉萨"]`，又能像 NumPy 向量化运算 `s * 2`
- **集字典和数组的优点于一身**

**今天 Series 简单了解就行**——重点是 DataFrame。

### 1.3 DataFrame —— Python 里的 Excel

**DataFrame 像一张表格**：
- 多行多列
- 每列是一个 Series（共享行 index）
- 每列可以是不同的类型（字符串、数字、日期）
- 提供大量数据分析方法

````python
# 从字典创建 DataFrame —— 字典的 key 是列名
df = pd.DataFrame({
    "city": ["北京", "上海", "广州", "拉萨"],
    "latitude": [39.90, 31.23, 23.13, 29.65],
    "elevation": [43, 4, 21, 3650],
    "temp": [13.2, 17.1, 22.5, 8.6]
})
df    # Jupyter 里直接输入变量名就显示美观的表格
````

**🌟 这就是 DataFrame**：
- 像 Excel 一样的表格视图
- 左边的 0、1、2、3 是**行 index**（默认从 0 开始）
- 顶上的 city、latitude... 是**列名**
- 每列可以是不同类型

## 四、选择数据 —— 列、行、条件筛选

**这是 Pandas 最重要的内容！**

### 4.1 选择一列

````python
df = pd.read_csv("data/cities_climate.csv")

# 用 df['列名'] 选一列
df["city"]    # 返回一个 Series
````

````python
# 直接对一列做统计
print("平均温度：", df["annual_temp"].mean())
print("最高温度：", df["annual_temp"].max())
print("最低温度：", df["annual_temp"].min())
print("温度标准差：", df["annual_temp"].std())
````

### 4.2 选择多列

````python
# 用 df[['列1', '列2']] 选多列（注意双方括号！）
df[["city", "annual_temp", "annual_rainfall"]]
````

**⚠️ 高频踩坑**：
- 选**一列**：`df["city"]` —— 单方括号、字符串
- 选**多列**：`df[["city", "temp"]]` —— **双方括号**、列表

**为什么**：
- `df["city"]` 中的字符串是单个列名
- `df[[...]]` 中的列表是多个列名 → 双方括号

### 4.3 选择行：iloc 和 loc

````python
# iloc —— 按整数位置（i = integer，类似 NumPy）
df.iloc[0]        # 第 0 行（返回 Series）
````

````python
# 切片：前 3 行
df.iloc[0:3]
````

````python
# 同时选行和列：iloc[行, 列]
df.iloc[0:3, 0:2]    # 前 3 行 + 前 2 列
````

````python
# loc —— 按标签（label）
# 当 index 是默认数字时，loc 和 iloc 表面看一样
df.loc[0]            # 标签为 0 的行
````

**📌 iloc vs loc 的核心区别**：

| 用法 | iloc | loc |
|---|---|---|
| 索引方式 | 按**整数位置** | 按**标签** |
| 切片 `0:3` | 0,1,2（不含3）| 0,1,2,3（**含 3**！）|
| 何时使用 | 数据按位置选 | 数据按标签选 |

**📌 经验法则**：
- **新手默认用 `iloc`**——和 Python/NumPy 的索引行为一致
- **当 index 是有意义的标签**（如城市名、日期）时，用 `loc` 更直观

### 4.4 ⭐⭐⭐ 条件筛选（布尔索引）

**Pandas 的核心特性 —— 像 SQL 一样筛选数据**！

````python
# 单条件筛选：年降水 > 1000
df[df["annual_rainfall"] > 1000]
````

**🌟 解读 `df[df["annual_rainfall"] > 1000]`**：

1. `df["annual_rainfall"] > 1000` —— 返回一个 True/False 的 Series
2. `df[True/False Series]` —— 返回 True 位置的所有行
3. **整体效果**：筛出符合条件的行

````python
# 复合条件：温度 > 15 且 降水 > 800
# ⚠️ 必须用 & 和括号（和 NumPy 一样！）
df[(df["annual_temp"] > 15) & (df["annual_rainfall"] > 800)]
````

````python
# 或：温度 < 10 或 降水 > 1500
df[(df["annual_temp"] < 10) | (df["annual_rainfall"] > 1500)]
````

**🛡️ 复合条件的两个铁律**（与 NumPy 一致）：

1. **必须用 `&` `|` `~`**（不能用 and/or/not！）
2. **每个条件必须用括号 `()`**（运算符优先级问题）

```python
# ❌ 错的写法
df[df["temp"] > 15 and df["rain"] > 800]    # 报错！

# ❌ 错的写法（缺括号）
df[df["temp"] > 15 & df["rain"] > 800]    # 优先级错乱

# ✅ 正确的写法
df[(df["temp"] > 15) & (df["rain"] > 800)]
```

### 4.5 isin —— 多值筛选

````python
# 想筛选'北京'、'上海'、'广州'三个城市
target_cities = ["北京", "上海", "广州"]
df[df["city"].isin(target_cities)]
````

**`isin()` 是 Pandas 的实用神技**——避免写一长串 `or` 条件。

### 6.2 缺失值（NaN）—— 真实数据的常态

**真实数据经常有缺失值**——传感器故障、用户没填、数据丢失等。

**Pandas 用 `NaN`（Not a Number）表示缺失值**。

````python
# 读取一个含缺失值的文件
df_messy = pd.read_csv("data/stations_with_nan.csv")
df_messy
````

**看到了吗？**——表格里有 `NaN`——这就是缺失值。

````python
# 检查每列有多少缺失值
df_messy.isna().sum()
````

````python
# 方法 1：删除含缺失值的行（dropna）
df_clean = df_messy.dropna()
df_clean
````

````python
# 方法 2：用值填充缺失（fillna）
df_filled = df_messy.fillna(0)    # 用 0 填
df_filled
````

````python
# 实用：用每列的均值填充
df_filled2 = df_messy.copy()
df_filled2["temperature"] = df_filled2["temperature"].fillna(df_messy["temperature"].mean())
df_filled2["rainfall"] = df_filled2["rainfall"].fillna(df_messy["rainfall"].mean())
df_filled2
````

**📌 缺失值处理的两种基本策略**：

1. **dropna()** —— 删除含缺失值的行（数据多时可用）
2. **fillna(值)** —— 填充缺失值（用 0 / 均值 / 中位数等）

**实战中怎么选**：
- 缺失少（< 5%）→ dropna
- 缺失多 → fillna（用合理的值）
- 关键变量缺失 → 必须 fillna 或专门处理

````python
# ===== 即时练习③ 参考答案 =====
import pandas as pd
import numpy as np

# 步骤1：读取
df = pd.read_csv("data/cities_climate.csv")

# 步骤2：comfort_score
df["comfort_score"] = (
    100 
    - abs(df["annual_temp"] - 20) * 5 
    - abs(df["annual_rainfall"] - 800) * 0.05
)

# 步骤3：climate_zone
df["climate_zone"] = np.where(df["annual_temp"] > 20, "热带",
                     np.where(df["annual_temp"] > 10, "温带", "寒带"))

# 步骤4：按 comfort_score 降序
df_sorted = df.sort_values("comfort_score", ascending=False)
print("=== 按舒适度降序 ===")
print(df_sorted[["city", "annual_temp", "annual_rainfall", "comfort_score", "climate_zone"]])
````

````python
# 步骤5：筛选 comfort_score > 60
comfortable = df_sorted[df_sorted["comfort_score"] > 60]
print(f"\n=== 舒适度 > 60 的城市（{len(comfortable)} 个）===")
print(comfortable[["city", "comfort_score", "climate_zone"]])

# 步骤6：climate_zone 分布
print("\n=== 各 climate_zone 城市数 ===")
print(df["climate_zone"].value_counts())

# 步骤7：写入 CSV
comfortable.to_csv("output_comfort.csv", index=False, encoding="utf-8-sig")
print("\noutput_comfort.csv 已写入！")
````

````python
# ===== 课后作业 2 参考答案 =====
import pandas as pd
import numpy as np

df = pd.read_csv("data/cities_climate.csv")

# 1. is_humid
df["is_humid"] = df["annual_rainfall"] > 1000
print("=== is_humid 添加后 ===")
print(df[["city", "annual_rainfall", "is_humid"]].head())

# 2. lat_zone
df["lat_zone"] = np.where(df["latitude"] >= 30, "North", "South")
print("\n=== lat_zone 添加后 ===")
print(df[["city", "latitude", "lat_zone"]].head())

# 3. isin 筛选
target = ["北京", "上海", "广州", "哈尔滨"]
selected = df[df["city"].isin(target)]
print("\n=== 4 个城市 ===")
print(selected[["city", "annual_temp", "is_humid", "lat_zone"]])

# 4. lat_zone 分布
print("\n=== lat_zone 分布 ===")
print(df["lat_zone"].value_counts())
````

````python
# ===== 课后作业 3 参考答案 =====
import pandas as pd
import numpy as np

df = pd.read_csv("data/stations_with_nan.csv")

# 1. 检查缺失
print("=== 缺失值统计 ===")
print(df.isna().sum())

# 2. 用均值填充
df["temperature"] = df["temperature"].fillna(df["temperature"].mean())
df["rainfall"] = df["rainfall"].fillna(df["rainfall"].mean())
print("\n=== 填充后 ===")
print(df.isna().sum())

# 3. 添加 temp_class
df["temp_class"] = np.where(df["temperature"] > 15, "warm",
                   np.where(df["temperature"] > 5, "mild", "cold"))
print("\n=== 添加 temp_class 后 ===")
print(df.head(8))

# 4. 写入
df.to_csv("output_filled_stations.csv", index=False, encoding="utf-8-sig")
print("\noutput_filled_stations.csv 已写入！")
````
