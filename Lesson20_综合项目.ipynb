### 方向 A：区域气候变化分析 ⭐⭐

**情境**：分析中国某区域多年的气候变化趋势。

**核心问题**：
- 选定区域的温度/降水有什么时间趋势？
- 不同子区域之间有什么差异？
- 极端气候事件（最热/最冷/最干/最湿月）的空间分布如何？

**用到的技能**：
- 第 11、14、15 课：Pandas 读 CSV + groupby + agg
- 第 16、17 课：GeoPandas 读 shp + sjoin（站点 → 区域）+ dissolve
- 第 19 课：多图报告页 + 专业可视化

**推荐数据**：
- 中国气象数据网（http://data.cma.cn）：站点月/年值数据
- 或用国家气象科学数据中心的开放数据
- **简化版本**：用本节课提供的扩展 weather_records.csv（已含多年数据）

**分析步骤示例**：
1. 读取多年气象数据，做基础清洗
2. 按区域 / 年份 groupby 统计趋势
3. 用 sjoin 把站点关联到行政区划
4. 用 dissolve 合并到区域级别
5. 画时间序列图 + 空间分布图 + 综合报告页

## 二、项目模板演示 —— 一个完整的'最小可行项目'

**用 30 行代码演示一个完整项目**——你的项目要比这个**更丰富**。

**示例方向**：方向 A 的简化版（用课程数据集）

````python
# ============================================================
# 期末项目模板演示 —— 区域气候差异分析
# 作者：[你的姓名]
# 日期：2024年X月X日
# ============================================================

# 1. 导入与配置
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False

print('=== 期末项目示范 —— 区域气候差异分析 ===\n')
````

````python
# 2. 数据加载
print('【步骤 1】加载数据')

# 气象记录（84 条月度数据）
records = pd.read_csv('data/weather_records.csv')
print(f'  气象记录：{records.shape}')

# 观测站位置（含 region）
stations = gpd.read_file('data/weather_stations.shp')
print(f'  观测站：{len(stations)} 个')

# 行政区划
provinces = gpd.read_file('data/provinces_simple.shp')
print(f'  省份：{len(provinces)} 个')
````

````python
# 3. 数据分析（核心）
print('\n【步骤 2】数据分析')

# 3.1 各区域的年平均温度和总降水
region_stats = records.groupby('region').agg(
    avg_temp=('temperature', 'mean'),
    total_rain=('rainfall', 'sum'),
    n_records=('temperature', 'count')
).round(1)
print('\n各区域气候统计：')
print(region_stats)

# 3.2 温差最大的月份
monthly_range = records.groupby('month')['temperature'].agg(['max', 'min'])
monthly_range['range'] = monthly_range['max'] - monthly_range['min']
max_range_month = monthly_range['range'].idxmax()
print(f'\n温差最大的月份：{max_range_month} 月（温差 {monthly_range.loc[max_range_month, "range"]:.1f}°C）')
````

````python
# 4. 综合可视化报告页
print('\n【步骤 3】生成综合报告页')

fig, axes = plt.subplots(2, 2, figsize=(16, 11))

# 左上：观测站地图
ax_map = axes[0, 0]
provinces.plot(ax=ax_map, color='lightyellow', edgecolor='gray', linewidth=0.5)
stations.plot(ax=ax_map, color='red', markersize=100, marker='^',
              edgecolor='white', linewidth=1.5)
for idx, row in stations.iterrows():
    ax_map.annotate(row['city'], xy=(row.geometry.x, row.geometry.y),
                    xytext=(7, 7), textcoords='offset points',
                    fontsize=10, fontweight='bold',
                    bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8))
ax_map.set_title('观测站位置', fontsize=13, fontweight='bold')
ax_map.set_xlabel('经度'); ax_map.set_ylabel('纬度')
ax_map.grid(True, alpha=0.3, linestyle='--')

# 右上：区域年均温（柱状图）
ax_temp = axes[0, 1]
bars = ax_temp.bar(region_stats.index, region_stats['avg_temp'],
                    color=['#3498DB', '#2ECC71', '#E74C3C', '#F39C12'], edgecolor='black')
for bar in bars:
    height = bar.get_height()
    ax_temp.text(bar.get_x() + bar.get_width()/2, height + 0.5,
                  f'{height:.1f}°C', ha='center', fontsize=11)
ax_temp.set_title('各区域年平均温度', fontsize=13, fontweight='bold')
ax_temp.set_ylabel('°C')
ax_temp.grid(True, alpha=0.3, axis='y')

# 左下：各月温度范围（折线 + 填充）
ax_range = axes[1, 0]
ax_range.fill_between(monthly_range.index, monthly_range['min'], monthly_range['max'],
                      alpha=0.3, color='#3498DB', label='温度范围')
ax_range.plot(monthly_range.index, monthly_range['max'], marker='o', color='#E74C3C', label='最高')
ax_range.plot(monthly_range.index, monthly_range['min'], marker='o', color='#3498DB', label='最低')
ax_range.set_title('各月温度范围（全国最高/最低对比）', fontsize=13, fontweight='bold')
ax_range.set_xlabel('月份'); ax_range.set_ylabel('°C')
ax_range.set_xticks(range(1, 13))
ax_range.legend(); ax_range.grid(True, alpha=0.3)

# 右下：区域降水箱线
ax_rain = axes[1, 1]
regions = records['region'].unique()
data_box = [records[records['region']==r]['rainfall'].values for r in regions]
bp = ax_rain.boxplot(data_box, labels=regions, patch_artist=True, widths=0.6)
for patch, color in zip(bp['boxes'], ['#3498DB', '#2ECC71', '#E74C3C', '#F39C12']):
    patch.set_facecolor(color); patch.set_alpha(0.6)
ax_rain.set_title('各区域月降水分布', fontsize=13, fontweight='bold')
ax_rain.set_ylabel('降水 (mm)')
ax_rain.grid(True, alpha=0.3, axis='y')

fig.suptitle('2024 年中国部分区域气候综合分析', fontsize=17, fontweight='bold', y=1.00)
plt.tight_layout()

# 5. 保存高清图
plt.savefig('output_demo_report.png', dpi=300, bbox_inches='tight')
plt.show()

print('\n✅ 报告页已保存为 output_demo_report.png')
````

````python
# 6. 结论与发现
print('\n【步骤 4】结论与发现')
print('='*50)

# 自动生成结论文字
warmest_region = region_stats['avg_temp'].idxmax()
wettest_region = region_stats['total_rain'].idxmax()

print(f'1. 最温暖区域：{warmest_region}（年均温 {region_stats.loc[warmest_region, "avg_temp"]:.1f}°C）')
print(f'2. 降水最多区域：{wettest_region}（总降水 {region_stats.loc[wettest_region, "total_rain"]:.0f} mm）')
print(f'3. 温差最大月份：{max_range_month} 月（最高与最低相差 {monthly_range.loc[max_range_month, "range"]:.1f}°C）')
print(f'4. 全国温度差异显著：1月{monthly_range.loc[1, "range"]:.1f}°C，7月{monthly_range.loc[7, "range"]:.1f}°C')

print('\n=== 项目演示完成 ===')
````
